```python
## 모듈 로딩
import pymysql

print(f"pymysql v.{pymysql.__version__}")
```

```python
## RDBMS 정보
SEVER_IP = 'localhost' 
USER_ID  = 'root'
USER_PW  = 'rlagusdn@0'
DB_NAME  = 'mydb'
CHARSET  = 'uft8mb4'
```

```python
## ================================================================
## DB 연결
## ================================================================
conn = pymysql.connect(host        = SEVER_IP,
                       user        = USER_ID,
                       password    = USER_PW,
                       database    = DB_NAME,
                       cursorclass = pymysql.cursors.DictCursor)
```

```python
## ================================================================
## DB - SELECT 처리
## ================================================================
ALL_DATA_SQL = 'SELECT * FROM user;'
ID_DATA_SQL  = 'SELECT id FROM user'

with conn.cursor() as cursor:
    cursor.execute(ALL_DATA_SQL)
    result = cursor.fetchall()  # 모든 데이터 가져오기
    display(result)
```

```python
## ================================================================
## DB - INSERT 처리
## ================================================================
INSERT_DATA_SQL = "INSERT INTO user VALUES (17,  '마징가', 23, '부산', '부산시', '여');"


with conn.cursor() as cursor:
    cursor.execute(INSERT_DATA_SQL)
    conn.commit()
```

```python
## ================================================================
## DB - DELETE 처리
## ================================================================
DELETE_DATA_SQL = "DELETE FROM user WHERE id=%s;"


with conn.cursor() as cursor:
    cursor.execute(DELETE_DATA_SQL, (11,))
    conn.commit()
```

```python
## ================================================================
## DB - UPDATE 처리
## ================================================================
UPDATE_DATA_SQL = "UPDATE user SET name=%s WHERE id=%s;"


with conn.cursor() as cursor:
    cursor.execute(UPDATE_DATA_SQL, ('원더우먼', 10))
    conn.commit()
```

```python
## ================================================================
## DB 연결 해제
## ================================================================
conn.close()
```
