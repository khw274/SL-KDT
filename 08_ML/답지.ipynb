## III. 주관식 정답 (모범 답안)

### 1. K-Culture의 성공요인 3가지
**정답:**
- **크리에이터-시청자 간의 개방적 관계**: 글로벌 OTT의 등장으로 기존의 폐쇄적 크리에이터-시청자 관계가 개방적/다원적 관계로 전환되었고, 한국 시청자들이 능동적 크리에이터(Storyteller)로 변화했음.
- **다양성과 융합**: 한국의 압축적 성장과정에서 경험한 다양한 사회·문화적 변화(산업화, 민주화, 글로벌화 등)를 바탕으로 다양한 컨텐츠를 제작할 수 있는 능력.
- **체계적인 선순환 생태계**: 투자 조합, 제작사, 유통사, 인재 육성 기관이 유기적으로 연결되어 프리미엄 콘텐츠를 지속적으로 생산하는 시스템.

### 2. OTT 시대의 콘텐츠 소비 변화
**정답:**
- 시간 제약 및 공간 제약의 해소로 주도적 시청이 가능해짐
- 롱폼 + 규격화 + 심의로부터의 자유
- 롱, 숏, 미드 폼의 다양한 길이의 콘텐츠 제공
- 타깃팅이 명확해져 특정 시청층에 맞춘 콘텐츠 제작 가능
- Binge watching(한 번에 몰아보기) 시청 패턴 활성화
- 국경 파괴로 글로벌 콘텐츠 접근이 용이해짐

### 3. 음악 콘텐츠의 시대별 변화 흐름
**정답:**
- **1960년대 후반~1970년대**: 라디오 음악 프로그램으로 시작 (KBS '밤을 잊은 그대에게', MBC '별이 빛나는 밤에')
- **1970년대 후반~1980년대**: 대학가요제 등장으로 아마추어 등용문 확대 (MBC 대학가요제, 강변가요제)
- **1990년대**: 기획사 중심의 아티스트 발굴 열풍 (뮤직뱅크, 인기가요)
- **2000년대~2010년대**: 서바이벌/오디션 프로그램의 확산 (슈퍼스타K, 나는 가수다, K팝스타)
- **2010년대 후반~2020년대**: 특정 타깃과 장르별 프로그램 다양화 (미스트롯, 싱어게인, 스트릿 우먼 파이터)

### 4. 뉴스 매체의 다원화가 일어난 배경과 사례
**정답:**
- **배경**: 디지털 시대의 진입으로 뉴스 소비 방식 변화, 시청자의 다양한 정보 추구 요구, 케이블 채널 활성화
- **사례**:
  - 보도 전문 채널: YTN(1995년), 연합뉴스TV(2011년)
  - 종합 편성 채널: JTBC, MBN, TV조선, 채널A(모두 2011년 12월 개국)
  - 디지털 플랫폼: 유튜브 저널리즘 활성화로 지상파/종편의 유튜브 채널 개설

### 5. 스포츠 방송의 진화 과정과 매체 변화
**정답:**
- **1927년**: 전조선 야구선수권 쟁탈전 중계로 최초 스포츠 프로그램 시작 (청각 매체 중심)
- **1960년대**: TV 3사의 스포츠 중계 경쟁 시작 (흑백 TV)
- **1980년대**: 88 서울 올림픽을 계기로 아날로그 칼라 TV 중계차 및 제작 시스템 운영
- **1995년**: 스포츠 전문 케이블 채널(SBS SPORTS) 개국
- **2008년**: 프로야구 4개 전 구장 TV, 인터넷 동시 생중계 시작
- **현재**: OTT 플랫폼의 독점 중계 (쿠팡플레이, 스포티비 등)

### 6. 먹방과 쿡방이 결합된 이유와 특징

  - 호기심 및 오감 유발형 콘텐츠
  - 요리 정보와 대결 포맷 활용
  - 실시간 사운드(리얼사운드) 강조
  - 다양한 연령층 타깃팅

- **유행 배경**:

### 9. 다큐멘터리의 형식 변화 (시대순)
**정답:**
- **1920년대**: 로버트 플래허티의 '북극의 나누크' - 내러티브를 가진 최초의 다큐멘터리
- **1930년대**: 지가 베르토프의 '편찬 영화' - 소비에트 실험주의 다큐
- **1960년대**:
  - 다이렉트 시네마 (미국): 객관적 관찰, 내레이션 없음
  - 시네마 베리테 (유럽): 인터뷰/토론 포함, 감독의 적극적 참여
- **1980년대**: 성찰적 양식 - 제작자의 주관적 입장 드러냄, 마이클 무어 스타일
- **2000년대 이후**: 모큐멘터리, 인터렉티브 다큐, 애니메이션 다큐, VR 다큐 등 다양한 형식 등장

### 10. 휴먼 다큐멘터리의 특징과 최근 트렌드
**정답:**
- **특징**:

- **최근 트렌드**:
  - 기존의 높은 제작비 구조에서 벗어나 기동성 강조
  - 짧은 형식의 다큐 콘텐츠 제작
  - 이웃의 일상적 관심사(건강, 취미, 특기) 중심 소재 선택
  - 현장성 강조, 실험적 구성 시도

### 11. 개그 프로그램의 시대별 특징 변화
**정답:**
- **1960~1970년대**: 슬랩스틱 코미디 중심 (육체적, 행동중심)
- **1980년대**: 정치 풍자 코미디 시대 (성대모사, 비판)

- **2000년대 이후**:
  - 버라이어티와 결합 (SNL KOREA, 코미디 빅리그)
  - 시청자 참여 강화 (댓글 소개, 청중 참여)
  - 포맷 다양화 (리그제, 매주 달라지는 호스트)
- **최근**: 정치 풍자 축소, 자유로운 표현 재정의 필요

### 12. 유튜브 콘텐츠의 성공 전략 3가지
**정답:**
- **전략 1: Profitability (수익성)**
  - 조회수가 많아도 수익이 나지 않으면 의미 없음
  - PPL, 협찬, 부가 수익 확보 필수
  - 제작비와 수익의 균형 고려
  
- **전략 2: Identity (정체성)**
  - 전문가들이 유튜브로 진입하는 시장에서 차별화 필요
  - 자신만의 고유한 콘텐츠 정체성 구축
  - 경쟁 콘텐츠와의 차별점 명확화
  
- **전략 3: Content Analysis & Market Analysis (분석력)**
  - SWOT 분석으로 자신의 콘텐츠 객관적 평가
  - 마케팅 4P(광고주 관점), 4C(시청자 관점) 전략 활용
  - 경쟁 콘텐츠, 시청자, 시장 환경 분석

### 13. ChatGPT의 학습 방식 3가지
**정답:**
- **Instruction Following (지시 따르기)**
  - 사용자의 지시사항을 이해하고 따르는 능력
  - 명확한 프롬프트 입력 시 원하는 답변 도출 용이
  
- **In-Context Learning (문맥 이해)**
  - 입력된 데이터의 문맥을 이해하여 답변 생성
  - 이전 대화의 맥락을 유지하며 응답
  - 시뮬레이션이나 데이터 기반 학습 가능
  
- **Step-by-step Reasoning (단계별 추론)**
  - 복잡한 문제를 단계별로 분해하여 해결
  - Chain-of-thought 프롬프팅으로 논리적 사고 유도

### 14. 프롬프트 엔지니어링의 기초 원리 3가지
**정답:**
- **구체성 (명확성)**
  - AI가 구체적인 프롬프트를 받으면 원하는 키워드와 내용에 대한 확률이 높아짐
  - 모호한 표현보다 구체적인 지시사항이 더 좋은 결과 도출
  
- **데이터 & In-context Learning (문맥 이해)**
  - AI는 입력된 데이터를 기반으로 학습하고 문맥을 이해하여 답변 생성
  - 원하는 답변이 AI의 데이터베이스에 존재하는지 확인 필요
  - 예시 제공이 중요
  
- **알고리즘 (심플화)**
  - 아무리 구체적으로 작성해도 한꺼번에 너무 많은 지시를 하면 확률 저하
  - 문제해결을 먼저 알고리즘화하여 분석하고 심플하게 정리 필요

### 15. 저작권의 발생과 보호 기간
**정답:**
- **발생**:
  - 저작물을 창작한 시점부터 자동 발생 (무방식주의)
  - 별도의 등록, 납본, 형식 요구 사항 없음
  - 산업재산권과 달리 출원/등록이 불필요
  
- **보호 기간**:
  - 원칙: 저작자가 생존하는 동안과 사망한 후 70년간
  - 공동저작물: 맨 마지막 저작자가 사망한 후 70년간
  - 영상저작물: 공표된 때로부터 70년간
  - 보호 기간 계산: 저작자 사망 또는 공표 연도의 다음해 1월 1일부터 시작

### 16. 저작인격권과 저작재산권의 차이
**정답:**
- **저작인격권** [제11조~15조]:
  - 저작자의 명예와 인격을 보호하기 위한 권리
  - 공표권: 저작물을 공표할 결정 권리
  - 성명표시권: 저작물의 원본, 복제물에 실명 또는 이명 표시 권리
  - 동일성유지권: 저작물의 내용, 형식, 제호의 변경을 허락/금지할 권리
  - 양도 불가능 (저작자 본인만 갖는 권리)
  
- **저작재산권** [제16조~22조]:
  - 저작물을 배타적으로 이용할 수 있는 권리
  - 복제권, 공중송신권, 배포권, 공연권, 전시권, 대여권, 2차적저작물작성권
  - 양도 가능 (제3자에게 판매/이전 가능)
  - 경제적 이익 추구 목적

### 17. 뉴스기사 저작권과 일반 사진의 저작권 귀속
**정답:**
- **뉴스기사의 저작권**:
  - 저작권자: 기사를 작성한 기자가 아닌, 기자의 소속 언론사
  - 이유: 업무상 저작물로서 법인, 단체, 사용자를 저작자로 인정 (저작권법 제9조)
  - 뉴스 기사 사용 시 반드시 해당 언론사의 허락 필수
  
- **일반 사진의 저작권**:
  - 저작권자: 사진을 촬영한 사진사/사진가 본인
  - 예외: 회사에 고용된 카메라맨이 업무상 촬영한 사진은 회사가 저작권 보유 가능

### 18. 무방식주의와 산업재산권의 개념 비교
**정답:**
- **무방식주의 (저작권법)**:
  - 저작권은 별도의 절차나 형식(납본, 등록 등) 없이 자동 발생
  - 창작물이 만들어지는 순간 보호됨
  - 국제 조약 (베른 협약)에 의한 원칙
  - 저작권은 등록과 무관하게 존재
  
- **산업재산권 (특허법, 상표법, 디자인법)**:
  - 특허청에 출원하여 등록되어야만 권리 발생
  - 등록 전에는 권리가 존재하지 않음
  - 산업 및 경제활동과 관련된 발명, 브랜드 등을 보호
  - 등록 후에만 독점적 권리 행사 가능

### 19. 현대 미디어 환경에서 K-콘텐츠의 지속 가능성을 위한 전략
**정답:**
- **수익 분배 구조 개선**:
  - 넷플릭스 같은 글로벌 OTT와의 공정한 계약 필수
  - 창작자/제작사에 대한 공정한 수익 배분 시스템 구축
  - 오징어게임의 사례처럼 지적재산권 보유 필요
  
- **K-OTT의 국제 경쟁력 강화**:
  - 정부의 적극적 투자 및 지원
  - K-OTT 서비스의 글로벌 확장 (현재 중화권 36%, 동남아 18.5%, 일본 15.4%)
  
- **법제도 정비**:
  - 재영상 분배금 같은 재사용 권리의 법적 보호
  - 이상한 변호사 우영우처럼 방영권만 판매하고 지적재산권 유지 사례 확대
  
- **콘텐츠 다양성 추구**:
  - 단순 수익성만 추구하지 않고 문화적 가치 고려
  - 독립 제작사와 창작자들에 대한 지원 강화

### 20. 디지털 시대 콘텐츠 제작 시 고려해야 할 법적, 윤리적 문제점
**정답:**
- **저작권 관련 법적 이슈**:
  - 뉴스기사 사용 시 반드시 언론사 허락 필수 (저작권법)
  - 사진, 음악, 동영상 등 타인 저작물 무단 사용 금지
  - 2차적저작물작성권 침해에 대한 규제
  
- **저작권 보호 제외 사항 숙지**:
  - 헌법, 법률, 판결, 행정 문서 등은 저작권 대상 아님
  - 부고, 인사동정, 주식시세는 사실 전달에 불과한 시사보도로 보호 제외
  - 이러한 내용을 자유롭게 사용 가능
  
- **디지털 콘텐츠 제작 시 주의사항**:
  - AI 생성 콘텐츠와 인간 작성 콘텐츠의 구분 명시 필요
  - 타인의 초상권, 영상사용권, 음향권 등 다양한 권리 침해 가능성 검토
  - 가짜뉴스(Misinformation, Disinformation) 생성 금지
  
- **윤리적 이슈**:
  - 개인정보보호 (GDPR 등 규제 강화)
  - 표현의 자유와 인권침해의 균형
  - 다양성과 포용성 고려 (혐오 표현, 차별 금지)
  - 콘텐츠 투명성 (광고, PPL 표시)
  - 미성년자 보호 및 유해 콘텐츠 규제
  
- **AI 관련 이슈**:
  - AI 생성 콘텐츠 탐지 및 표시 (워터마크, 라벨링)
  - 학습 데이터의 출처와 저작권 확인
  - 인공지능의 불공정성 및 편향성 검토
