##### 정밀도와 재현율
- 쉽게 강아지 사진으로 예시를 들어보겠슴니다 

- 정밀도 TP / (FP + TP) 
    -(ex) 100개의 강아지 사진에서 A라는 모델이 40개를 강아지라고 했고 그중에서 36개가 진짜 강아지라면 -> precision = 90%    
- 재현율 TP / (FN + TP)
    -(ex) 100개의 강아지 사진에서 B라는 모델이 40개를 강아지라고 맞게 찾아낸 정답 -> 40/100 -> Recall : 40%  

- 예시로 쉽게 이해하기

- 정밀도와 재현율 지표 중에 이진 분류 모델의 업무 특성에 따라 특정 평가 지표가 더 중요한 지표로 간주될수도 있음!!
    - **재현율이 중요 지표인 경우 실제 P 양성 데이터를 N으로 잘못 판단하게 된다면 업무상 큰 영향이 발생**

##### 정밀도/재현율 트레이드오프
- 정밀도와 재현율은 상호 보완적인 평가 지표이기에 어느 한 쪽을 강제로 높이면 다른 하나의 수치는 떨어지기 쉽다
    - → 이를 **정밀도/재현율의 트레이드오프(Trade-off)라고 부른다!!**

- 임계값(Threshold)를 조정해 정밀도 or 재현율의 수치를 높일 수 있다
    - 이것도 굉장히 쉽게 예시를 들어보자면~ 
        - 어떤 이미지의 "강아지일 확률" = 0.65 라면 threshold = 0.7 -> 이다 그럼 강아지 아님(0)
        threshold = 0.5 -> 강아지 맞음 (1)
    
    - 정밀도를 올리고 싶다! -> threshold 높이기
        - 정말 확실한 경우에만 강아지라고 true
    - 재현율을 올리고 싶다! -> threshold 낮추기 
        - 조금만 강아지처럼 보여도 true

    - 즉 임계값은 모델 확률에 대한 기준선이라고 보면됨 

쉽게 생각해보자 ex) iris data

OVR (One vs Rest)
- setosa     vs (versicolor + virginica)
- versicolor vs (setosa + virginica)
- virginica  vs (setosa + versicolor)

최종 예측 -> 각 모델이 출력하는 decision score중 가장 높은 모델의 클래스로 선택
- 장점 !
    - -> 클래스가 많아도 학습 부담이 적다 
- 단점 !!
    - -> 클래스 간 경계가 애매하면 정확도가 떨어질 수 있다
    - -> 불균형 데이터에서 편향이 생길수도 있음 




OvO (One vs One)
- setosa     vs versicolor
- setosa     vs virginica
- versicolor vs virginica

**총 3개의 모델을 학습함 이게 지금은 작아서 적어보이는데 갯수가 많아지면 경우의수도 그만큼 늘어난다고 보면됨**

- 장점 !
    - -> 각 모델이 두 종류 클래스만 처리 -> 경계가 더 명확해질수도 있다
    - -> 비선형 모델에서 성능이 좋은 경우가 많다

- 단점 ! 
    - -> 클래스 개수가 많아지면 모델 수가 매우 매우 매우 많아진다
    - -> ex) 100개 클래스면.. 4950개의 모델이 필요한거지
