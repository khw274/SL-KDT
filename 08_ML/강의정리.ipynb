## 1주차: K-Culture 현황

### K-Culture의 범위
- **K-Pop**: 음악
- **K-Drama**: 드라마
- **K-Movie**: 영화
- **K-Food**: 음식
- **K-Life**: 라이프스타일

### 한국 드라마의 성공 사례

**Pre K-Culture**
- **사랑이 뭐길래(1991)**: MBC 평균 시청률 59.6%
- **겨울연가(2002)**: 중국 CCTV 채널 1 평균 시청률 4.2% (약 1억 5천만 명 시청)
- **대장금(2003)**: 약 60여 개국 수출, 약 3조원 경제적 효과

### K-Pop의 성공

**강남스타일(2012) - 싸이**
- YouTube 창시 이래 최초 161일 만에 10억 뷰 달성

**BTS(방탄소년단)**
- 공식 YouTube 채널 구독자 7,710만 명 (2023년 12월)
- 빌보드 핫 100 1위 데뷔 (2020년 6월 27일)
- 가수 최초로 UN 총회 연설

**BLACKPINK(블랙핑크)**
- 공식 YouTube 채널 구독자 9,250만 명 (2023년 12월)
- 한국 걸그룹 최초 빌보드 200 1위 달성 (2022년 10월 1일)

### K-Movie의 성공

**기생충(2020) - 봉준호 감독**
- 제 92회 아카데미: 작품상, 감독상, 각본상, 국제영화상

**미나리(2021) - 정이삭 감독**
- 제 93회 아카데미: 여우조연상

**헤어질 결심(2022) - 박찬욱 감독**
- 제 75회 칸 영화제 감독상

**오징어 게임(2021)**
- 제 74회 에미상 13개 부문 노미네이트, 6관왕 달성

**이상한 변호사 우영우**
- 넷플릭스 비영어권 드라마 부문 글로벌 시청시간 1위 (7주 연속)

### 콘텐츠 산업 성장도

**부가가치 창출 효과(2020년 기준)**
- 반도체: 79조 원
- 콘텐츠: 51조 원
- 자동차: 47조 원

**수출액 추이(2016~2021)**
- 2016년: 60.1억 달러
- 2021년: 124.5억 달러 (연평균 15.7% 증가)

**한류의 경제 효과**
- 한류를 접한 후 한국 상품 구매 경험: 75% (2011년 기준)

**화장품 산업(세계 시장, 2020년 기준)**
- 1위: 프랑스 (75억 7천만 달러)
- 2위: 미국 (10억 6천만 달러)
- 3위: 독일 (3억 4천만 달러)
- 4위: 한국

## 2주차: OTT 환경

### OTT(Over The Top)의 정의
- **구독 경제**: 방송 콘텐츠, 영화 등 각종 콘텐츠를 스트리밍 서비스로 제공

### 2023년 국내 OTT 서비스 월간 활성 사용자(MAU)

| 순위 | 서비스 | 사용자 수 | 특징 |
|------|--------|---------|------|
| 1 | Netflix | 664만 | 글로벌 탑티어, 한국 2023년 1,137만 명 |
| 2 | TVING | 562만 | CJ ENM/KT 시즌 합병 |
| 3 | 쿠팡플레이 | 519만 | 와우 멤버십, 로켓배송 연동 |
| 4 | WAVVE | 474만 | KBS/MBC/SBS/SK텔레콤 합작 |
| 5 | Disney+ | 339만 | 디즈니 콘텐츠 독점 |
| 6 | 왓챠 | 81만 | - |

**OTT 가입자 수 변화**
- 2016년: 456만 명
- 2023년: 3,008만 명 (약 6.6배 증가)

### Netflix 분석

**특징**
- 기존 서비스를 온라인으로 옮긴 글로벌 탑티어
- 다양한 자체 콘텐츠 개발 및 투자
- 초기부터 자본 투자로 5개년 안정적 순이익 구조 구축

**오리지널 K-콘텐츠**
- '스위트홈'
- '경성크리처'
- '이두나!'

**2022년 실적(전년대비)**
- 수익: 316.16억 달러 (↑6.46%)
- 운영비: 68.14억 달러 (↑10.43%)
- 순이익: 44.92억 달러 (↓-12.20%)

### TVING(티빙)

**개요**
- 대한민국 토종 OTT 서비스
- CJ ENM 계열에서 출발, KT의 시즌(Seezn)과 합병

**지분 구성**
- CJ ENM: 48.85%
- 네이버, KT스튜디오지니 등: 10% 내외

**2023년 12월 기준**
- 가입자: 약 510만 (성장 정체)
- 최근 CJENM 수익성 악화로 티빙-웨이브 합병설

### WAVVE(웨이브)

**개요**
- 지상파 3사(KBS, MBC, SBS)와 SK텔레콤 합작
- 2019년 9월 출범
- 기존 'POOQ', 'oksusu' 등 흡수

**콘텐츠**
- 영화, 드라마, 해외시리즈, 야구, 오리지널, 클래식

**2023년 12월 기준**
- 가입자: 약 422만 (성장 정체)
- 3년 연속 적자 (SK그룹 지원에도 불구하고)

### 쿠팡플레이

**특징**
- 2020년 12월 출시
- 유료 멤버십(와우)으로 서비스
- 로켓배송과 추가 콘텐츠 이용 연동
- 수익모델에서 다소 자유로움 (아마존 프라임 전략 벤치마킹)

**킬러 콘텐츠**
- 영화, TV시리즈
- 스포츠 중계
- SNL 코리아 시즌 4
- 미국 NFL

### 디즈니플러스

**특징**
- 2019년 4월 론칭
- 막강한 콘텐츠와 자본력을 가진 거대 기업 OTT
- Netflix에서 마블 영화 축소

**오리지널 작품**
- '무빙' 등

## 3주차: 음악 프로그램 및 콘텐츠

### 시대별 음악 프로그램

#### 1960년대 후반 ~ 1970년대: 음악 콘텐츠 시작

**특징**
- 생방송 라이브의 묘미
- 라디오 방송 프로그램

**주요 프로그램**
- **KBS 밤을 잊은 그대에게** (1964년 5월 9일~현재)
  - 1964년 편성되어 현재까지 방송 중인 최장수 프로그램
  - 편안한 음악과 따뜻한 이야기로 힐링

- **MBC 별이 빛나는 밤에** (1969년 3월 17일~현재)
  - 1969년 편성되어 현재까지 방송 중
  - MBC 표준 FM 음악 전문 라디오 프로그램

#### 1970년대 후반~1980년대: 대학가요제의 등장

**배경**: 대학 입학이 쉽지 않아 대학생들을 대상으로 한 프로그램 등장

**주요 프로그램**
- **MBC 대학가요제** (1977년 9월 12일~2012년 11월 8일)
  - MBC에서 매년 주최한 대학생 대상 가요제

- **MBC 강변가요제** (1979년 10월 1일~2001년 11월 30일)
  - MBC가 매년 여름 북한강 강변에서 열던 가요 경연대회
  - 대학가요제와 함께 1980~'90년대 중반 가요제 양대 산맥

- **KBS2 가요톱 10** (1981년 2월 10일~1998년 2월 11일)
  - KBS 2TV에서 방송한 대중음악 순위 프로그램 (현 뮤직뱅크 전신)

#### 1990년대: 기획사 중심 아티스트 발굴 열풍

**주요 프로그램**
- **KBS 뮤직뱅크** (1998년 1월 16일~현재)
  - 1990년대 KBS에서 방영된 예능 프로그램 중 유일하게 종영하지 않고 꾸준히 방영

- **SBS 인기가요** (2000년 2월 1일~현재)
  - 매주 일요일 오후 3시 40분 방송
  - 생방송 가요 순위 프로그램

#### 2000년대~2010년대: 서바이벌 프로그램 열풍

**악동클럽: 꼴지탈출 시즌 3** (2001년 5월 26일~2002년 2월 24일)
- MBC의 '목표달성! 토요일' 한 코너
- 아이돌 그룹 선발 오디션
- 전국 고등학생 대상

**Mnet 슈퍼스타 K** (2009년 7월 24일~2016년 12월 8일)
- 케이블 방송 엠넷의 대국민 공개 오디션 프로그램
- 2016년까지 시즌이 이어짐

**MBC 나는 가수다** (2011년 11월 3일~2015년 4월 24일)
- 가수들이 노래를 불러 청중평가단에게 심사받는 서바이벌
- 2015 시즌 3까지 제작

**SBS KPOP STAR** (2011년 12월 4일~2017년 4월 9일)
- 차세대 K팝 스타를 발굴하는 서바이벌 프로그램
- 실시간 투표 시스템이 화제
- 방송사 오디션 프로그램에서 시청자 선택 강화 (참여성 증가)

### 음악 콘텐츠 트렌드 분석

#### 버추얼 아이돌

**특징**
- 제목, 이미지, 음악이 어우러져 듣는 콘텐츠를 넘어 스토리텔링이 더해진 음악 콘텐츠

**주요 사례**
- **왁타버스** (구독자 65.3만 명)
- **플레이브(PLAVE)** (구독자 54.5만 명)
  - 아이돌 같은 세계관
  - 유튜브 채널을 통한 꾸준한 라이브 소통
  - 커버 곡으로 아이돌처럼 활동
  - 음악방송, 행사 등에 출연해 버추얼 아이돌의 입상 증명

- 지상파: 주요 경기만 중계

**특징**

4. **광고주**: 광고료 집행

- 콘텐츠 제작 → 플랫폼 → 구독료/광고료 → 방송사

- **GPT-2** (2019년 2월 14일): 15억 개 (12.8배 ↑)

- **GPT-3.5** (2022년 11월 30일): 1,750억 개 (116.7배 ↑ from GPT-1)

#### 학습 방식

1. **Instruction Following**
   - 입력 프롬프트 바탕으로 최적의 답변 생성

2. **In-Context Learning**
   - 수만 개의 데이터셋 추가 학습
   - 질문-답변 형식 지도학습

3. **Step-by-step Reasoning**
   - 인간 피드백을 통한 강화학습(RLHF) 적용

#### GPT-3의 한계 (Hallucination)

**문제**: 없는 정보를 마치 있는 것처럼 제시

**예시**
- Q: 조선왕조 실록에 기록된 세종의 맥북 던지기 사건
- A (GPT-3): 마치 있는 것처럼 답변
- A (GPT-3.5): 조선왕조 실록은 역사서이며, 맥북 같은 현대 기술은 존재하지 않음을 정정

## 15주차: 저작권

### 저작물과 저작권

#### 저작물의 정의

**저작물** (저작권법 제2조 제1호)
- 인간의 사상 또는 감정을 표현한 창작물

#### 사상과 감정의 해석

**학설 1**: 사상이란 사고의 내용을 지칭
**학설 2**: 철학적, 심리적 개념으로 좁게 해석하지 않음 (넓은 의미)
**학설 3**: '생각보다 더 넓은 의미'로 봐야 함

### 저작권 보호 예외

#### 보호 받지 못하는 저작물

다음에 해당하는 것은 저작권법의 보호를 받지 못함:

1. **헌법, 법률, 조약, 명령, 조례 및 규칙**
2. **국가 또는 지방자치단체의 고시, 공고, 훈련** 등
3. **법원의 판결, 결정, 명령** 및 행정심판 절차 결과
4. **국가 또는 지방자치단체가 작성한 것**으로 위 1~3호 규정의 편집물 또는 번역물
5. **사실의 전달에 불과한 시사보도** (≠뉴스저작물)

### 저작권 관련 법률

#### 저작권자의 권리 (제9, 10조)

**저작자의 권리**
- 저작인격권
- 저작재산권
- 예외: 업무상 저작물에 한하여 법인, 단체, 사용자를 저작자로 인정

#### 저작인격권 (제11조~15조)

**목적**: 저작자의 명예와 인격 보호

**구체적 권리**

1. **공표권**: 저작물을 공표할지 결정하는 권리
2. **성명표시권**: 저작물의 원본, 복제물, 공표 매체 등에 실명 또는 이명 표시 권리
3. **동일성유지권**: 저작물의 내용, 형식, 제호의 변경 허락/금지 권리

#### 저작재산권 (제16조~22조)

**목적**: 저작물을 배타적으로 재배할 수 있는 권리

**구체적 권리**

1. **복제권**: 저작물 복제 권리
2. **공중송신권**: 유무선통신 송신이나 이용 제공 권리
3. **배포권**: 저작물 원본이나 복제물 배포 권리
4. **공연권**: 저작물 공연 권리
5. **전시권**: 원본이나 복제물 전시 권리
6. **대여권**: 상업용 음반, 프로그램을 영리 목적으로 대여할 권리
7. **2차적저작물작성권**: 번역, 편곡, 변형, 각색, 영상제작 권리

### 저작권의 발생 (제10조)

**특징**: **무방식주의**
- 저작권은 저작물을 창작한 때부터 발생
- 어떠한 절차나 형식 (등록, 납본 등)을 요구하지 않음
- 산업재산권(특허권 등)과는 다름

### 저작권 보호 기간

**변화 추이**
- 1957년: 30년
- 1986년: 50년
- 2011년: 70년

**현행 보호 기간**

1. **일반 저작물**
   - 저작자가 생존하는 동안과 사망한 후 70년간 존속

2. **공동저작물**
   - 맨 마지막 사망한 저작자가 사망한 후 70년간 존속

3. **영상저작물**
   - 공표된 때로부터 70년

**보호 기간 계산**
- 저작자가 사망하거나 저작물을 공표한 해의 다음해 1월 1일부터 시작

### 디지털 콘텐츠 제작 시 주의사항

#### 저작물의 종류

1. **신문, 인터넷의 뉴스**: 어문 제작물
2. **언론사 기자 촬영 보도사진**: 사진 저작물
3. **방송뉴스 기자 보도**: 무형의 구술에 의한 어문 저작물
4. **방송뉴스의 음향 및 음악**: 음악 저작물
5. **방송, 인터넷 영상 뉴스**: 영상 저작물

#### 중요 원칙

**단순 사실 보도는 저작권법의 보호 대상이 아님**

#### 뉴스기사 사용 시 주의

**필수**: 언론사의 허락을 받아야 함
- **이유**: 뉴스 기사의 저작권은 기자가 아닌 **기자의 소속 언론사**가 보유

#### 보호 대상이 아닌 뉴스 기사

**사실 전달에 불과한 시사보도** (제7조 제5호)
- 부고 (訃告)
- 인사동정
- 주식시세
- 일기예보
- 6하 원칙에 따라 작성된 사건 사고 단신 기사
