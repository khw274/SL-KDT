```python
## ëª¨ë“ˆ ë¡œë”©
import numpy as np
import cv2
import matplotlib.pyplot as plt 
import koreanize_matplotlib
import os
```

```python
DATA_DIR = './hw'
SAVE_DIR = './hw/roi'


os.makedirs(SAVE_DIR, exist_ok=True)


files = sorted(os.listdir(DATA_DIR))


terminate = False  # ì „ì²´ í”„ë¡œê·¸ë¨ ì¢…ë£Œ í”Œë˜ê·¸


for idx, filename in enumerate(files):
    
    if terminate:
        break
    
    img_path = os.path.join(DATA_DIR, filename)
    img = cv2.imread(img_path)
    
    if img is None:
        print("ë¡œë“œ ì‹¤íŒ¨:", filename)
        continue
    
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    roi_selected = False
    
    while not roi_selected:
        # 1ï¸âƒ£ ì´ë¯¸ì§€ í‘œì‹œ
        print(f"\n[{idx+1}/{len(files)}] {filename}")
        print("ë“œë˜ê·¸ë¡œ ROI ì„ íƒ | [ESC] ì¦‰ì‹œ ì¢…ë£Œ")
        
        # 2ï¸âƒ£ ROI ì„ íƒ (ë§ˆìš°ìŠ¤)
        roi = cv2.selectROI(
            "ROI Selection",
            img_rgb,
            showCrosshair=True,
            fromCenter=False
        )
        
        x, y, w, h = roi
        
        # ESC í‚¤ ê°ì§€ (ë“œë˜ê·¸ ì•ˆ í•œ ìƒíƒœì—ì„œ ESC)
        if w == 0 and h == 0:
            print("í”„ë¡œê·¸ë¨ ì¢…ë£Œ")
            cv2.destroyAllWindows()
            terminate = True
            break
        
        # ì„ íƒëœ ROI í‘œì‹œ
        roi_img = img_rgb[y:y+h, x:x+w]
        
        # ROI ë¯¸ë¦¬ë³´ê¸°
        display_img = img_rgb.copy()
        cv2.rectangle(display_img, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.imshow("Preview", display_img)
        cv2.imshow("ROI", roi_img)
        
        print("ROI ì„ íƒë¨ | [SPACE] ì €ì¥ í›„ ë‹¤ìŒ | [R] ë‹¤ì‹œ ì„ íƒ | [ESC] ì¦‰ì‹œ ì¢…ë£Œ")
        
        # 3ï¸âƒ£ í‚¤ ì…ë ¥ ëŒ€ê¸°
        key = cv2.waitKey(0) & 0xFF
        
        # ğŸ”¹ SPACE: ì €ì¥ í›„ ë‹¤ìŒ ì´ë¯¸ì§€
        if key == 32:  # SPACE
            # íŒŒì¼ëª…ê³¼ í™•ì¥ì ë¶„ë¦¬
            name, ext = os.path.splitext(filename)
            # _roi ë¶™ì—¬ì„œ ì €ì¥
            roi_filename = f"{name}_roi{ext}"
            save_path = os.path.join(SAVE_DIR, roi_filename)
            cv2.imwrite(save_path, cv2.cvtColor(roi_img, cv2.COLOR_RGB2BGR))
            print(f"ğŸ’¾ ì €ì¥ ì™„ë£Œ: {save_path}")
            roi_selected = True
        
        # ğŸ”¹ ESC: ê°•ì œ ì¢…ë£Œ (ì €ì¥ ì•ˆ í•¨)
        elif key == 27:  # ESC
            print("í”„ë¡œê·¸ë¨ ì¢…ë£Œ")
            cv2.destroyAllWindows()
            terminate = True
            break
        
        cv2.destroyAllWindows()


cv2.destroyAllWindows()
print("\nROI ì„ íƒ ì¢…ë£Œ")
```
