#### 【 이미지 픽셀 분포 분석 - 밝고 어두움/명도 Intensity 】

- 밝기(Intensity) : 해당 픽셀이 얼마나 많은 빛 에너지를 가지는가를 나타냄.
    * R, G, B => 색 채널
    * 밝기 => 채널들의 종합 효과
- 히스토그램은 밝다/어둡다가 아닌 무슨 색이 많다를 보여주는 것 

[1] 모듈 로딩 및 데이터 준비<hr>

```python
## 모듈로딩
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 이미지 데이터 설정
IMG_FILE = '../Data/Image/mountain.jpg'


## 이미지 데이터 로딩
img_color = cv2.imread(IMG_FILE)
img_gray  = cv2.imread(IMG_FILE, cv2.IMREAD_GRAYSCALE)

print(img_gray.shape, img_gray.shape[0] * img_gray.shape[1])
```

[2] Grayscale 밝기 히스토그램 시각화<hr> 
- 인간 시각 가중 밝기 즉, 사람이 느끼는 밝기
- 변환식: Y=0.299R+0.587G+0.114B

```python
## ----------------------------------------------------
## Grayscale 밝기 히스토그램 
## ----------------------------------------------------
plt.figure(figsize=(6,4))
plt.hist(img_gray.ravel(), bins=256, range=(0,256))
plt.title("Intensity (Grayscale) Histogram")
plt.xlabel("Intensity")
plt.ylabel("Pixel Count")
plt.show()
```

[3] HSV V 채널 히스토그램 시각화<hr>
- 색+밝기 분리하여 가장 강한 색 채널 하나만 반영
- 조명 조건이 중요한 문제에 적합 
- 계산식 : V=max(R,G,B)

```python
## ---------------------------------------------------
## HSV에서 V 채널 분리 
## ---------------------------------------------------
hsv = cv2.cvtColor(img_color, cv2.COLOR_BGR2HSV)
v   = hsv[:,:,2]

plt.figure(figsize=(6,4))
plt.hist(v.ravel(), bins=256, range=(0,256))
plt.title("Brightness Histogram (HSV - V)")
plt.xlabel("Value")
plt.ylabel("Pixel Count")
plt.show()
```
