```python
# -------------------------------------------------------------------------
# 1-1. 모듈 로딩 
# -------------------------------------------------------------------------
import cv2
import os
import pandas as pd

# -------------------------------------------------------------------------
# 1-2. 데이터 로딩
# -------------------------------------------------------------------------
# 데이터 설정
DATA_DIR1 = ''
DATA_DIR2 = './bananas/img'

# 이미지 파일 리스트
FILES1 = os.listdir(DATA_DIR1)
FILES2 = os.listdir(DATA_DIR2)
```

```python
# -------------------------------------------------------------------------
# 사과 데이터 확인
# -------------------------------------------------------------------------
for idx, filename in enumerate(FILES1):
    if idx < 10:
        print(filename)
    else:
        break

# -------------------------------------------------------------------------
# 바나나 데이터 확인
# -------------------------------------------------------------------------
for idx, filename in enumerate(FILES2):
    if idx < 10:
        print(filename)
    else:
        break
```

[2] 기계학습을 위한 데이터셋 파일 생성 <hr>
- csv 파일 저장 => [형식] 타겟, 픽셀1, 픽셀2,........픽셀n
- 이미지 사이즈 일치
- 이미지 로우데이터 읽어오기
- 타겟 추출
- csv 파일에 쓰기

```python
# -------------------------------------------------------------------------
# 2-1. csv 파일명
# -------------------------------------------------------------------------
CSV_DIR = './apples/csv'
CSV_FILE = './apples/csv/fruits.csv'

if not os.path.exists(CSV_DIR1):
    os.mkdir(CSV_DIR1)
else:
    print(f'{CSV_DIR1} 경로가 이미 존재합니다.')
```

```python
# -------------------------------------------------------------------------
# 2-2. 이미지 처리
# -------------------------------------------------------------------------
# 사과 이미지 처리
# -------------------------------------------------------------------------
with open(CSV_FILE1, mode='w', encoding='utf-8') as f:
    
    # 파일 1개 처리
    for file_name in FILES1:
        img_path1 = f'{DATA_DIR1}/{file_name}'
        save_path1 = f'{DATA_DIR1}/re_{file_name}'
        print(img_path1)
        
        img1 = cv2.resize(cv2.imread(img_path1, cv2.IMREAD_GRAYSCALE),
                        (0, 0), 
                        fx=0.5, fy=0.5, 
                        interpolation=cv2.INTER_AREA)
        print('1차 :', img1.shape)
        # print(img, '\n')

        img1 = cv2.resize(img1, 
                        (70, 70), 
                        interpolation=cv2.INTER_AREA)
        print('2차 :', img1.shape)

        cv2.imwrite(save_path1, img1)

        img1 = img1.reshape(-1)
        print('3차(2D → 1D) :', img1.shape, '\n')

        raw = [str(i) for i in img1.tolist()]
        raw = ','.join(raw)
        # print(raw)

        f.write('apple,'+raw+'\n')
# -------------------------------------------------------------------------
# 바나나 이미지 처리
# -------------------------------------------------------------------------
with open(CSV_FILE2, mode='w', encoding='utf-8') as f:
    
    # 파일 1개 처리
    for file_name in FILES2:
        img_path2 = f'{DATA_DIR2}/{file_name}'
        save_path2 = f'{DATA_DIR2}/re_{file_name}'
        print(img_path2)
        
        img2 = cv2.resize(cv2.imread(img_path2, cv2.IMREAD_GRAYSCALE),
                        (0, 0), 
                        fx=0.5, fy=0.5, 
                        interpolation=cv2.INTER_AREA)
        print('1차 :', img2.shape)
        # print(img, '\n')

        img2 = cv2.resize(img2, 
                        (70, 70), 
                        interpolation=cv2.INTER_AREA)
        print('2차 :', img2.shape)

        cv2.imwrite(save_path2, img2)

        img2 = img2.reshape(-1)
        print('3차(2D → 1D) :', img2.shape, '\n')

        raw = [str(i) for i in img2.tolist()]
        raw = ','.join(raw)
        # print(raw)

        f.write('apple,'+raw+'\n')
```
