### [이미지 연산]
- 픽셀(pixel): 언산 수행 이미지 변화
    - 이미지 데이터 => 행열, numpy의 ndarray
    - 덧___셈 연산 => 각각의 픽셀 단위 덧셈
    - 뺄___셈 연산 => 각각의 픽셀 단위 뺄셈
    - 곱___셈 연산 => 각각의 픽셀 단위 곱셈
    - 나눗셈 연산 => 각각의 픽셀 단위 나눗셈

- 색상값 : 0 검정색 black - 1 또는 255 흰색 white
    - 0 >>>>>>>>>>> 1 또는 255 : 밝아짐
    - 0 <<<<<<<<<<< 1 또는 255 : 어두워짐
- 이미지 밝기 조절
- 이미지 자연스러운 합성

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# ==============================================
# 모듈 로딩
# ==============================================
import numpy as np
import cv2
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 데이터 파일
FILE_NAME = '../Data/Image/flower.jpg'
```

```python
# ==============================================
# 데이터 로딩 및 기본 정보 출력
# ==============================================
imgNP = cv2.imread(FILE_NAME)
grayNP = cv2.imread(FILE_NAME, cv2.IMREAD_GRAYSCALE)

print(f'imgNP: {imgNP.shape}, grayNP: {grayNP.shape}') 
```

[2] 이미지 밝기 조절 <hr>

```python
# ==============================================
# [2-1] 컬러 이미지 밝기 조절
# ==============================================
# 0 >>>>> 255 : 흰색, white 밝아짐
colorNP30 = cv2.add(imgNP, 30)
colorNP50 = cv2.add(imgNP, 50)
colorNP100 = cv2.add(imgNP, 100)

# ==============================================
# 시각화
# ==============================================
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
numOBJ = [imgNP, colorNP30, colorNP50, colorNP100]
titles = ['ORI', '+30', '+50', '+100']

for ax, img, title in zip(axes, numOBJ, titles):
    ax.imshow(img)
    ax.set_title(title)

plt.show()
```

```python
# 0 >>>>> 255 : 검은색, black 어두워짐
colorNP30 = cv2.subtract(imgNP, 30)
colorNP50 = cv2.subtract(imgNP, 50)
colorNP100 = cv2.subtract(imgNP, 100)
# ==============================================
# 시각화
# ==============================================
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
numOBJ = [imgNP, colorNP30, colorNP50, colorNP100]
titles = ['ORI', '+30', '+50', '+100']

for ax, img, title in zip(axes, numOBJ, titles):
    ax.imshow(img)
    ax.set_title(title)

plt.show()
```

```python
# 0 >>>>> 255 : 흰색, white 밝아짐
grayNP30 = cv2.add(grayNP, 30)
grayNP50 = cv2.add(grayNP, 50)
grayNP100 = cv2.add(grayNP, 100)

# ==============================================
# 
# 
# 시각화
# ==============================================
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
numOBJ = [imgNP, grayNP30, grayNP50, grayNP100]
titles = ['ORI', '+30', '+50', '+100']

for ax, img, title in zip(axes, numOBJ, titles):
    ax.imshow(img)
    ax.set_title(title)

plt.show()
```
