####  [OpenCV - 이미지 색공간(Color Space) 변경]

- 목적에 따른 색상 표현해주는 색공간 변경을 ㅜ이한 것
    - 함수: cv2.cvtColor(로우이미지데이터, code)
        * code 매개변수 : cv2.COLOR_BGR2RGB, cv2.COLOR_BGR, CV2.COLOR_BGR2HSV, ....
    - 결과: 새로운 색공간 적용된 이미지 로우 데이터

- **주의 : opencv 한글 미지원! 경로, 파일명 한글 사용 xxx**

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# 이미지 처리 모듈
import cv2

# 시각화 모듈
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# 이미지 파일
IMG_FILE = '../img.png'
```

[2] 이미지 로딩 및 정보 확인 <hr>

```python
ORG_IMG = cv2.imread(IMG_FILE)  # 컬러
GRAY_IMG = cv2.imread(IMG_FILE, cv2.IMREAD_GRAYSCALE)
```

```python
# 정보 확인
print(f'[원본 이미지 데이터]\n 형태: {ORG_IMG.shape}, 차원: {ORG_IMG.ndim}, 사이즈: {ORG_IMG.size}')
print(f'[그레이스케일 이미지 데이터]\n 형태: {GRAY_IMG.shape}, 차원: {GRAY_IMG.ndim}, 사이즈: {GRAY_IMG.size}')
```

```python
# 이미지 시각화
fig, axes = plt.subplots(1, 2, sharey=True)

# 컬러 이미지
axes[0].imshow(ORG_IMG)
axes[0].set_title('컬러 이미지')
axes[0].set_xticks([])   
axes[0].set_yticks([])  

# 그레이스케일 이미지
axes[1].imshow(GRAY_IMG, cmap='gray') 
axes[1].set_title('그레이 스케일 이미지')
axes[1].set_xticks([])
axes[1].set_yticks([])

plt.suptitle("cat")
plt.tight_layout()
plt.show()
```

[3] 이미지 색공간 분리 <HR>


```python
# =====================================================
# [3-1] BGR => RGB 색공간
# =====================================================
RGB_IMG = cv2.cvtColor(ORG_IMG, cv2.COLOR_BGR2RGB)

# 이미지 시각화
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(10, 5))

# 컬러 이미지
axes[0].imshow(ORG_IMG)
axes[0].set_title('컬러 이미지')
axes[0].set_xticks([])   
axes[0].set_yticks([])  

# RGB 이미지
axes[1].imshow(RGB_IMG)
axes[1].set_title('RGB 이미지')
axes[1].set_xticks([])
axes[1].set_yticks([])

plt.suptitle("BGR vs RGB")
plt.tight_layout()
plt.show()
```

```python
# =====================================================
# [3-2] BGR => HSV 색공간
# =====================================================
HSV_IMG = cv2.cvtColor(ORG_IMG, cv2.COLOR_BGR2HSV)  # ORG_IMG 배열 사용

# 이미지 시각화
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(10, 5))

# 컬러 이미지
axes[0].imshow(ORG_IMG)
axes[0].set_title('컬러 이미지')
axes[0].set_xticks([])   
axes[0].set_yticks([])  

# HSV 이미지
axes[1].imshow(HSV_IMG)
axes[1].set_title('HSV 이미지')
axes[1].set_xticks([])
axes[1].set_yticks([])

plt.suptitle("BGR vs HSV")
plt.tight_layout()
plt.show()
```

```python
# =====================================================
# [3-3] BGR => YCrCb 색공간
# =====================================================
YCrCb_IMG = cv2.cvtColor(ORG_IMG, cv2.COLOR_BGR2YCrCb)  # ORG_IMG 배열 사용

# 이미지 시각화
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(10, 5))

# 컬러 이미지
axes[0].imshow(ORG_IMG)
axes[0].set_title('컬러 이미지')
axes[0].set_xticks([])   
axes[0].set_yticks([])  

# HSV 이미지
axes[1].imshow(YCrCb_IMG)
axes[1].set_title('YCrCb 이미지')
axes[1].set_xticks([])
axes[1].set_yticks([])

plt.suptitle("BGR vs YCrCb")
plt.tight_layout()
plt.show()
```
