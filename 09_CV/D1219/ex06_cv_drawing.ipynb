####  [OpenCV - 이미지와 그리기 함수들]

- 객체 인식 시 객체 위치 지정용 사용됨
- 일반 그리기용으로도 사용됨

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# 이미지 처리 모듈
import cv2

# 시각화 모듈
import matplotlib.pyplot as plt
import koreanize_matplotlib

import numpy as np
```

```python
# 데이터 준비 : 도화지랑 이미지
IMG_FILE = '../Data/Image/lion_face.jpg'
PAPER_FILE = '../Data/Image/paper.jpg'
```

```python
# 흰색 도화지 이미지 생성 및 저장
ret = cv2.imwrite(PAPER_FILE, np.full((500, 500), 255, dtype=np.uint8))

print(f"save ok? {ret}")
```

```python
# 흰색 도화지 이미지 데이터 & 사진 이미지 데이터 로딩
PAPER_IMG = cv2.imread(PAPER_FILE)
IMG = cv2.imread(IMG_FILE)

# 정보 확인
print(f'[원본 이미지 데이터]\n 형태: {PAPER_IMG.shape}, 차원: {PAPER_IMG.ndim}, 사이즈: {PAPER_IMG.size}')
```

[2] 이미지 그리기 <hr>

```python
# [2-1] 직선 그리기
cv2.line(PAPER_IMG, 
                 pt1=(0, 0), pt2=(300, 300), 
                 color=(0, 255, 0), 
                 thickness=3)  # 원본 이미지에 직접 그리기
cv2.imshow('line', PAPER_IMG)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# [2-2] 사각형 그리기
cv2.rectangle(PAPER_IMG, 
                           (0, 0), (300, 300), 
                           (255, 0, 0), 
                           thickness=3)
cv2.imshow('rectangle', PAPER_IMG)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# [2-3] 다각형 그리기
points = np.array([[50,50], [150,50], [100,150]], np.int32)
cv2.polylines(PAPER_IMG, [points], 
                           isClosed=True, 
                           color=(255,0,0), 
                           thickness=3)
cv2.imshow('polyline', PAPER_IMG)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# [2-4] 원 그리기
cv2.circle(PAPER_IMG, (60, 60), 20, (0, 0, 255), thickness=3)
cv2.imshow('circle', PAPER_IMG)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# [2-5] 글자 쓰기
cv2.putText(PAPER_IMG, 'HANDSOME GUY HIM CHAN', (30, 30), cv2.FONT_HERSHEY_SIMPLEX, 
            fontScale=1, color=(255, 0, 0), thickness=2)
cv2.imshow('text', PAPER_IMG)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
