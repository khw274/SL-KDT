####  [OpenCV - 이미지 크기 변경]

- 이미지의 확대/축소 기능
    - 함수: cv2.cvtColor(로우이미지데이터, dsSize=(w, h), fx=상대크기, fy=상대크기, interporation=보간법)
    - 결과: 변경된 크기의 이미지 로우 데이터

- **주의 : opencv 한글 미지원! 경로, 파일명 한글 사용 xxx**

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# 이미지 처리 모듈
import cv2

# 시각화 모듈
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# 이미지 파일
IMG_FILE = '../Data/Image/field.bmp'
```

[2] 이미지 로딩 및 정보 확인 <hr>

```python
ORG_IMG = cv2.imread(IMG_FILE)  # 컬러
GRAY_IMG = cv2.imread(IMG_FILE, cv2.IMREAD_GRAYSCALE)
```

```python
# 정보 확인
print(f'[원본 이미지 데이터]\n 형태: {ORG_IMG.shape}, 차원: {ORG_IMG.ndim}, 사이즈: {ORG_IMG.size}')
print(f'[그레이스케일 이미지 데이터]\n 형태: {GRAY_IMG.shape}, 차원: {GRAY_IMG.ndim}, 사이즈: {GRAY_IMG.size}')
```

```python
# 이미지 시각화
fig, axes = plt.subplots(1, 2, sharey=True)

# 컬러 이미지
axes[0].imshow(ORG_IMG)
axes[0].set_title('컬러 이미지')
axes[0].set_xticks([])   
axes[0].set_yticks([])  

# 그레이스케일 이미지
axes[1].imshow(GRAY_IMG, cmap='gray') 
axes[1].set_title('그레이 스케일 이미지')
axes[1].set_xticks([])
axes[1].set_yticks([])

plt.suptitle("cat")
plt.tight_layout()
plt.show()
```

[3] 이미지 크기 <HR>


```python
# =====================================================
# [3-1] 절대 크기 변경 : 축소, 원본, 확대
# =====================================================
# 이미지 축소 : 1/3 크기
DOWN_IMG = cv2.resize(ORG_IMG, dsize=(87, 123), interpolation=cv2.INTER_AREA)

# 이미지 확대 : 1.5배
UP_IMG = cv2.resize(ORG_IMG, dsize=(392, 557), interpolation=cv2.INTER_CUBIC)

# 축소, 원본, 확대 이미지 시각화
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(cv2.cvtColor(DOWN_IMG, cv2.COLOR_BGR2RGB))
axes[0].set_title('축소 (1/3)')
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(ORG_IMG, cv2.COLOR_BGR2RGB))
axes[1].set_title('원본')
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(UP_IMG, cv2.COLOR_BGR2RGB))
axes[2].set_title('확대 (1.5배)')
axes[2].axis('off')

plt.suptitle("절대 크기 변경")
plt.tight_layout()
plt.show()
```

```python
# =====================================================
# [3-2] 상대 크기 변경 : 축소, 원본, 확대
# =====================================================
# 이미지 축소 : 0.3, 0.5 배
DOWN_03_IMG = cv2.resize(ORG_IMG, dsize=(0, 0), fx=0.3, fy=0.3, interpolation=cv2.INTER_AREA)
DOWN_05_IMG = cv2.resize(ORG_IMG, dsize=(0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)

# 이미지 확대 : 1.5, 2.5 배
UP_15_IMG = cv2.resize(ORG_IMG, dsize=(0, 0), fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)
UP_25_IMG = cv2.resize(ORG_IMG, dsize=(0, 0), fx=2.5, fy=2.5, interpolation=cv2.INTER_CUBIC)

# 축소, 원본, 확대 이미지 시각화
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

axes[0,0].imshow(cv2.cvtColor(DOWN_03_IMG, cv2.COLOR_BGR2RGB))
axes[0,0].set_title('축소 0.3배')
axes[0,0].axis('off')

axes[0,1].imshow(cv2.cvtColor(DOWN_05_IMG, cv2.COLOR_BGR2RGB))
axes[0,1].set_title('축소 0.5배')
axes[0,1].axis('off')

axes[0,2].imshow(cv2.cvtColor(ORG_IMG, cv2.COLOR_BGR2RGB))
axes[0,2].set_title('원본 (1.0배)')
axes[0,2].axis('off')

axes[1,0].imshow(cv2.cvtColor(UP_15_IMG, cv2.COLOR_BGR2RGB))
axes[1,0].set_title('확대 1.5배')
axes[1,0].axis('off')

axes[1,1].imshow(cv2.cvtColor(UP_25_IMG, cv2.COLOR_BGR2RGB))
axes[1,1].set_title('확대 2.5배')
axes[1,1].axis('off')

axes[1,2].axis('off')  # 빈 공간

plt.suptitle("상대 크기 변경 비교")
plt.tight_layout()
plt.show()
```
