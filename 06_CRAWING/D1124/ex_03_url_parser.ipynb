### 【 URL Parser 】
- URL주소를 분리/해석해 주는 모듈
    * http://www.naver.com/search/... => 6개 요소 분석
- urllib패키지내에 존재

[1] 모듈 로딩 <hr>

```python
## python 내장 패키지 중 url파싱 모듈
from urllib.parse import urljoin, urlparse
```

[2] ulr 파싱<hr>
- url 구성 : scheme://도메인+포트/패스;파라미터?쿼리#세션

- urlparse(url_str) 함수
    * 6개 필드를 가진 ParseResult 객체입니다.
    * scheme : 프로토콜 (http, https, ftp 등)
    * netloc : 도메인 + 포트 (ko.wikisource.org, example.com:8080)
    * path : /wiki/... 같은 경로
    * params : 오래된 스타일의 파라미터 (거의 안 씀)
    * query : ? 뒤의 쿼리스트링 (a=1&b=2)
    * fragment: # 뒤의 앵커 (section1)

```python
## URL 문자열 
url = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC?page=2&sort=title#section1"

## URL 문자열 파싱 => 6개요소로 해석 및 추출 가능 
parsed = urlparse(url)

print(parsed)
print("scheme :", parsed.scheme)    # https
print("netloc :", parsed.netloc)    # ko.wikisource.org
print("path   :", parsed.path)      # /wiki/저자:윤동주 (URL 인코딩 상태)
print("params :", parsed.params)    # 거의 안 씀
print("query  :", parsed.query)     # page=2&sort=title
print("fragment:", parsed.fragment) # section1
```

[3] URL 합치기<hr>

```python
## URL 연결 예 [1] ===============================

## - 베이스 URL
base = "http://example.com/html/a.html"

## - URL 조합
print( urljoin(base, "b.html") )
print( urljoin(base, "sub/c.html") )
print( urljoin(base, "../index.html") )
print( urljoin(base, "../img/hoge.png") )
print( urljoin(base, "../css/hoge.css") )
```

```python
## URL 연결 예[2] ===============================

## - 베이스 URL
base = "http://example.com/html/a.html"

## - URL 조합
print( urljoin(base, "/hoge.html") )
print( urljoin(base, "http://otherExample.com/wiki") )
print( urljoin(base, "//anotherExample.org/test") )
```
