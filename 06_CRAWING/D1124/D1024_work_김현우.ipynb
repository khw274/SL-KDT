```python
# 데이터 크롤링
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
```

```python
# 데이터 분석
import pandas as pd
```

[2] 크롤링할 위키문헌 사이트

```python
url = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC?page=2&sort=title#section1"
base_url = "https://ko.wikisource.org"  # 기본 도메인
```

```python
# 프로그램(봇) 요청이라서 접근을 거부함
# Request 객체에 User-Agent 헤더를 추가해 브라우저처럼 보이도록 함 
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})

bs = BeautifulSoup(urlopen(req), 'html.parser')

# === debugging ===
# print(bs.title)

# 1. 시 제목 추출 ======================================================================================================================================================
# 추출한 HTML 코드
# ex) 자화상
# <li>
#   <a href="/wiki/%ED%95%98%EB%8A%98%EA%B3%BC_%EB%B0%94%EB%9E%8C%EA%B3%BC_%EB%B3%84%EA%B3%BC_%EC%8B%9C_(1948%EB%85%84)/%EC%9E%90%ED%99%94%EC%83%81" 
#       title="하늘과 바람과 별과 시 (1948년)/자화상">자화상
#   </a>
# </li>


# ex) 소년
# <li>
#   <a href="/wiki/%ED%95%98%EB%8A%98%EA%B3%BC_%EB%B0%94%EB%9E%8C%EA%B3%BC_%EB%B3%84%EA%B3%BC_%EC%8B%9C_(1948%EB%85%84)/%EC%86%8C%EB%85%84" 
#       title="하늘과 바람과 별과 시 (1948년)/소년">소년
#   </a>
# </li>
# =====================================================================================================================================================================
titles = []
contents = []

# === debugging ===
# li_tag = bs.select('li')
# print(li_tag)

# *= ~ : ~가 포함된 요소 선택(부분 일치)
poem = bs.select('li > a[title*="하늘과 바람과 별과 시"]')
# === debugging ===
# print(poem)

# 2. 시 내용 추출 ======================================================================================================================================================
# 추출한 HTML 코드
# ex) 자화상
# <div class="poem">
#   <p>
#       "산모퉁이를 돌아 논가 외딴우물을 홀로 찾어가선 가만히 들여다봅니다."
#       <br>
#       <br>
#       "우물속에는 달이 밝고 구름이 흐르고 하늘이 펼치고 파아란 바람이 불고 가을이 있습니다."
#       <br>
#       <br>
#       "그리고 한 사나이가 있습니다."
#       <br>
#       "어쩐지 그 사나이가 미워저 돌아갑니다."
#       <br>
#       <br>
#       "돌아가다 생각하니 그 사나이가 가엾어집니다. 도로 가 들여다 보니 사나이는 그대로 있습니다.
#       <br>
#       <br>
#       "다시 그 사나이가 미워져 돌아갑니다."
#       <br>
#       돌아가다 생각하니 그 사나이가 그리워집니다."
#       <br>
#   </p>
#   <div class="__nop wst-nop"></div>
# </div>

# <div class="poem">
#   <p>우물속에는 달이 밝고 구름이 흐르고 하늘이 펼치고 파아란 바람이 불고 가을이 있고 追憶처럼 사나이가 있습니다.
#   </p>
# </div>
# =====================================================================================================================================================================

# 시 목록 리스트 (위 데이터처럼 'a' 태그 리스트가 있다고 가정)
poem = bs.select('li > a[title*="하늘과 바람과 별과 시"]')
# print(poem)

for a_tag in poem:
    # print(a_tag)
    titles.append(a_tag.get_text())  # 시 제목 텍스트 저장

    poem_url = a_tag['href']  # '/wiki/...' 형태 상대경로
    # === debugging ===
    # print(poem_url)
    
    poem_url = base_url + poem_url
    # === debugging ===
    # print(url)
    
    req = Request(poem_url, headers={'User-Agent': 'Mozilla/5.0'})
    bs = BeautifulSoup(urlopen(req), 'html.parser')
    
    content = bs.select('div.poem')
    # print(content)


    if content:
        poem_text = content[0].get_text(separator='\n').strip() 
        print(f"✓ {a_tag.get_text()}\n")
        # print(poem_text)
    else:
        poem_text = ""
        
    contents.append(poem_text)

# 데이터프레임 생성 ======================================================================================================================================================
datadf = pd.DataFrame({
    '시 제목': titles,
    '내용': contents
})

display(datadf.head()) 
# =====================================================================================================================================================================
# CSV 파일로 저장 (utf-8-sig 옵션은 한글 인코딩 문제 방지용)
datadf.to_csv('시집_내용.csv', index=False, encoding='utf-8-sig')

print("CSV 파일이 '시집_내용.csv' 이름으로 저장되었습니다.")
```
