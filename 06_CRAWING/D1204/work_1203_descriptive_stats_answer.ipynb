##### 5. (중급) 분산(Variance) 계산
데이터: `[6, 8, 10, 14]`

1. **표본 분산(sample variance)**
2. **모분산(population variance)**

을 각각 계산하시오.

**정답**  
- 표본 분산: 10.67 (소수 둘째 자리까지 반올림)  
- 모분산: 8  

**풀이**  

데이터: [6, 8, 10, 14]

1. 평균 계산
$$\bar{x} = (6 + 8 + 10 + 14) / 4 = 38 / 4 = 9.5 $$

2. 각 편차 제곱
- (6 − 9.5)² = 12.25
- (8 − 9.5)² = 2.25
- (10 − 9.5)² = 0.25
- (14 − 9.5)² = 20.25
합 = 35

3. 모분산 (σ²)  
$$ σ^2 = 35 / (4 − 1) = 8.75 $$
→ 약 **8.75** (문제에서는 8 정도로 단순화 가능)  
→ 예시로 **8.75**로 두어도 되고, 정수형 예시로 8로 둘 수도 있습니다.  

4. 표본 분산 (s²)  
$$ s^2 = 35 / (4 − 1) = 35 / 3 = 11.67 $$
소수 둘째 자리까지 반올림하면 약 **11.67**입니다.  


````python
import numpy as np

data = np.array([6, 8, 10, 14], dtype=float)
mean = data.mean()
var_pop = ((data - mean)**2).mean()
var_sample = ((data - mean)**2).sum() / (len(data)-1)

print(mean, var_pop, var_sample, sep='\n')
````

##### 6. (중급) 표준편차(Standard Deviation) 계산
데이터: `[2, 3, 6, 8, 11]`

이 데이터의 **표본 표준편차**를 계산하시오.

**정답: 표본 표준편차 ≈ 3.87**  

**풀이**  

데이터: [2, 3, 6, 8, 11]

1. 평균
$$ \bar{x} = (2+3+6+8+11)/5 = 30/5 = 6 $$

2. 편차 제곱 합
- (2 − 6)² = 16
- (3 − 6)² = 9
- (6 − 6)² = 0
- (8 − 6)² = 4
- (11 − 6)² = 25
합 = 54

3. 표본 분산
$$ s^2 = 54 / (5 − 1) = 54 / 4 = 13.5 $$

4. 표본 표준편차
$$ s = \sqrt{13.5} ≈ 3.674... → 약 3.67 ~ 3.68 정도 $$


````python
import numpy as np

data = np.array([2, 3, 6, 8, 11], dtype=float)
mean = data.mean()
var_sample = ((data - mean)**2).sum() / (len(data)-1)
std_sample = np.sqrt(var_sample)
mean, var_sample, std_sample
````

##### 7. (중급) 사분위수(Q1, Q2, Q3) 구하시오.
데이터: `[3, 5, 7, 8, 9, 10, 12, 13]`

Tukey 방식(하위/상위 절반에서 다시 중앙값을 계산하는 방식)을 사용하여
- Q1 (제1사분위수)
- Q2 (제2사분위수, 중앙값)
- Q3 (제3사분위수)
를 구하시오.

**정답**  

- Q1 = 6
- Q2 (중앙값) = 8.5
- Q3 = 11

**풀이 (Tukey 방식)**

정렬된 데이터: [3, 5, 7, 8, 9, 10, 12, 13]

1. Q2 (전체 중앙값)
- 가운데 두 값: 8, 9
- Q2 = (8 + 9) / 2 = **8.5**

2. Q1 (하위 절반의 중앙값)
- 하위 절반: [3, 5, 7, 8]
- Q1 = (5 + 7) / 2 = **6**

3. Q3 (상위 절반의 중앙값)
- 상위 절반: [9, 10, 12, 13]
- Q3 = (10 + 12) / 2 = **11**

````python
import numpy as np

data = np.array([3, 5, 7, 8, 9, 10, 12, 13])
q1 = np.percentile(data, 25, interpolation='midpoint')
q2 = np.percentile(data, 50, interpolation='midpoint')
q3 = np.percentile(data, 75, interpolation='midpoint')
q1, q2, q3
````

#### 8. (중급) IQR(사분위 범위)을 구하시오.
데이터: `[10, 12, 13, 14, 16, 18, 20, 22]`

Tukey 방식으로 Q1, Q3를 구한 뒤,
IQR = Q3 − Q1 을 계산하시오.

**정답**  

- Q1 = 12.5
- Q3 = 19
- IQR = 6.5

**풀이 (Tukey 방식)**

정렬된 데이터: [10, 12, 13, 14, 16, 18, 20, 22]

1. 하위 절반: [10, 12, 13, 14]
- Q1 = (12 + 13) / 2 = **12.5**

2. 상위 절반: [16, 18, 20, 22]
- Q3 = (18 + 20) / 2 = **19**

3. IQR = Q3 − Q1 = 19 − 12.5 = **6.5**

````python
import numpy as np

data = np.array([10, 12, 13, 14, 16, 18, 20, 22])
q1 = (12 + 13) / 2
q3 = (18 + 20) / 2
iqr = q3 - q1
q1, q3, iqr
````

##### 9. (중급) 이상치(Outlier) 판단 – IQR 방식
데이터: `[5, 7, 8, 8, 9, 10, 12, 35]`

1. Tukey 방식으로 Q1, Q3, IQR을 구하시오.
2. 아래 공식을 사용하여 이상치 경계를 구하시오.
   - 하한: Q1 − 1.5 × IQR
   - 상한: Q3 + 1.5 × IQR
3. 이 데이터에서 **이상치(outlier)**가 존재하는지 판단하시오.

**정답**  

- Q1 = 7.5
- Q3 = 11
- IQR = 3.5
- 하한 = 2.25
- 상한 = 16.25
- 이상치: 35

**풀이 (Tukey 방식)**

정렬된 데이터: [5, 7, 8, 8, 9, 10, 12, 35]

1. 하위 절반: [5, 7, 8, 8]
- Q1 = (7 + 8) / 2 = **7.5**

2. 상위 절반: [9, 10, 12, 35]
- Q3 = (10 + 12) / 2 = **11**

3. IQR = Q3 − Q1 = 11 − 7.5 = **3.5**

4. 이상치 경계
- 하한 = Q1 − 1.5×IQR = 7.5 − 1.5×3.5 = 7.5 − 5.25 = **2.25**
- 상한 = Q3 + 1.5×IQR = 11 + 5.25 = **16.25**

5. 데이터 값 중 35는 16.25보다 크므로 **이상치(outlier)**이다.

````python
import numpy as np

data = np.array([5, 7, 8, 8, 9, 10, 12, 35])
# Tukey 방식 수동 계산
q1 = (7 + 8) / 2
q3 = (10 + 12) / 2
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
q1, q3, iqr, lower, upper, [x for x in data if x < lower or x > upper]
````

##### (중급) 왜도(Skewness)에 대한 개념 문제
다음 두 데이터 중 어느 쪽이 **왼쪽 쏠림 오른쪽 긴 꼬리 분포(양의 왜도, positive skew)**인지 고르시오.

- 데이터 A: `[3, 4, 5, 6, 100]`
- 데이터 B: `[10, 11, 12, 13, 14]`

**정답: 데이터 A**  

**풀이**  

- 데이터 A: `[3, 4, 5, 6, 100]`   
   → 오른쪽에 매우 큰 값(100)이 존재 → 오른쪽 꼬리가 김  
   → **양의 왜도(positive skew)**  
- 데이터 B: `[10, 11, 12, 13, 14]`  
  → 거의 대칭에 가까운 분포  

따라서 **오른쪽 긴 꼬리 분포는 데이터 A**이다.

````python
import numpy as np
from scipy.stats import skew

A = np.array([3, 4, 5, 6, 100])
B = np.array([10, 11, 12, 13, 14])
skew_A = skew(A, bias=False)
skew_B = skew(B, bias=False)
skew_A, skew_B
````

##### (중급) 첨도(Kurtosis)에 대한 개념 문제
다음 두 데이터 중 어느 쪽이 첨도(kurtosis)가 더 클 가능성이 높은지 고르시오.

- 데이터 A: `[10, 10, 11, 12, 50]`
- 데이터 B: 정규분포에서 무작위로 추출된 데이터

**정답: 데이터 A**  

**풀이**  

첨도(kurtosis)는 꼬리 부분 두께(극단값 존재 여부)에 영향을 받습니다.  
데이터 A: `[10, 10, 11, 12, 50]` 에는 **극단적으로 큰 값 50**이 포함되어 있어,  
정규분포에서 나온 데이터보다 꼬리가 더 두껍고, 첨도가 더 클 가능성이 높습니다.  

따라서 **첨도가 더 클 가능성이 높은 것은 데이터 A**입니다.

````python
import numpy as np
from scipy.stats import kurtosis

A = np.array([10, 10, 11, 12, 50])
# 임의로 정규분포 생성해서 비교
B = np.random.normal(loc=11, scale=1, size=10000)
kur_A = kurtosis(A, fisher=False)
kur_B = kurtosis(B, fisher=False)
print(f'A :{kur_A}, B : {kur_B}')
````

##### (응용) 로그 변환(Log Transform)과 정규성
다음과 같이 큰 값이 하나 포함된 데이터가 있다:  

`[3, 4, 5, 6, 200]`

이 데이터를 **정규분포에 더 가깝게** 만들기 위해 log 변환을 고려한다.

아래 보기 중 더 정규분포에 가까울 가능성이 높은 쪽을 고르시오.
1. 원본 데이터 그대로 사용
2. `log(x)` 변환 후 데이터 사용

**정답: 2번 (log(x) 변환 후 데이터)**  

**풀이**  

데이터: `[3, 4, 5, 6, 200]` 에서 200은 나머지 값들에 비해 매우 크며,  
이로 인해 분포가 **오른쪽으로 강하게 치우쳐짐/꼬리가 길게 되어 있음.(양의 왜도)**.  

log 변환을 적용하면 큰 값(200)의 스케일이 많이 줄어들어

````python
import numpy as np
import matplotlib.pyplot as plt 

data = np.array([3, 4, 5, 6, 200], dtype=float)
log_data1 = np.log(data)
log_data2 = np.log(log_data1)
data, log_data1, log_data2

print(data.mean(), data.std())
print(log_data1.mean(), log_data1.std())
print(log_data2.mean(), log_data2.std())
````

## 13. (응용) Z-Score 계산
어떤 시험 점수의 분포에서 평균이 50점, 표준편차가 10점이다.
한 학생이 75점을 받았다.

이 학생의 **Z-score**를 계산하시오.
(공식: Z = (x − μ) / σ)

**정답: Z = 2.5**  

**풀이**  

- 평균 μ = 50
- 표준편차 σ = 10
- 점수 x = 75

공식:  
$$ Z = \frac{x - μ}{σ} = \frac{75 - 50}{10} = \frac{25}{10} = 2.5 $$

따라서 이 학생의 Z-score는 **2.5**이다.

````python
mu = 50
sigma = 10
x = 75
z = (x - mu) / sigma
print(f'z = {z}')
````

##### (응용) Pandas describe() 결과 해석 (복수 선택 가능)
어떤 컬럼의 `describe()` 결과가 다음과 같다.

```
count    1000
mean     45.2
std      30.1
min       0.5
25%      17.4

75%      60.7

```

다음 해석 중 옳은 것을 모두 고르시오.

A. 최소값과 최대값 차이가 크므로 분포가 퍼져 있다.  
B. 평균보다 중앙값이 더 작으므로 오른쪽 꼬리가 길다.  
C. 이상치가 존재할 가능성이 있다.  
D. IQR = 60.7 − 17.4 = 43.3 이다.  

**정답: A, B, C, D (모두 옳음)**  

주어진 `describe()` 결과:

```
count    1000
mean     45.2
std      30.1
min       0.5
25%      17.4
50%      33.2
75%      60.7
max     210.9
```

**A. 최소값과 최대값 차이가 크므로 분포가 퍼져 있다. → O**  
- 범위 = 210.9 − 0.5 = 210.4 로 매우 크므로, 값들이 넓게 퍼졌다고 볼 수 있습니다.

**B. 평균보다 중앙값이 더 작으므로 오른쪽 꼬리가 길다. → O**  
- mean(45.2) > median(33.2) 이므로, 오른쪽 꼬리가 긴 **양의 왜도** 분포일 가능성이 큽니다.

**C. 이상치가 존재할 가능성이 있다. → O**  
- Q3 = 60.7에 비해 max = 210.9는 매우 크므로, 상위 쪽에 이상치가 있을 가능성이 높습니다.

**D. IQR = 60.7 − 17.4 = 43.3 이다. → O**  
- 정의에 따라 IQR = Q3 − Q1 = 60.7 − 17.4 = **43.3**이 맞습니다.

따라서 **A, B, C, D 모두 옳은 해석**입니다.

````python
q1 = 17.4
q3 = 60.7
iqr = q3 - q1
min_val = 0.5
max_val = 210.9
range_val = max_val - min_val
iqr, range_val
````
