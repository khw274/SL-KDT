##### 1. 사건과 여사건
어떤 학생이 시험에서 **합격할 확률**이 0.7이다.  
이 학생이 **불합격할 확률(여사건)**을 구하시오.

**정답: 0.3**  
**풀이**  

어떤 사건 A(합격)의 확률이 P(A) = 0.7 이면,
그 여사건 Aᶜ(불합격)의 확률은

$$ P(A^c) = 1 - P(A) = 1 - 0.7 = 0.3 $$

따라서 불합격할 확률은 **0.3**이다.

```python
p_pass = 0.7
p_fail = 1 - p_pass
p_fail
```

##### 2.사건의 합집합 (Union)
한 마을에서
- 자전거를 가지고 있을 확률: 0.4  
- 킥보드를 가지고 있을 확률: 0.3  
- 둘 다 가지고 있을 확률: 0.15  

이 마을에서 임의의 한 사람을 선택했을 때,
**자전거나 킥보드 중 적어도 하나**는 가지고 있을 확률을 구하시오.

**정답: 0.55**  
**풀이**  
사건 A: 자전거를 가지고 있음  
사건 B: 킥보드를 가지고 있음

주어진 값:
- P(A) = 0.4
- P(B) = 0.3
- P(A ∩ B) = 0.15

적어도 하나는 가지고 있을 사건은 **A ∪ B** 이므로,

$$ P(A \cup B) = P(A) + P(B) - P(A \cap B)  $$
$$ = 0.4 + 0.3 - 0.15 = 0.55  $$

따라서 확률은 **0.55**이다.

```python
p_A = 0.4
p_B = 0.3
p_A_and_B = 0.15
p_union = p_A + p_B - p_A_and_B
p_union
```

##### 3. 사건의 교집합과 독립성
어떤 학생이 영어시험과 수학시험에 동시에 합격할 확률이 0.25이다.  
영어시험에 합격할 확률은 0.5, 수학시험에 합격할 확률은 0.4일 때,  
두 사건(영어 합격, 수학 합격)은 **독립적인지 여부**를 판단하시오.  

**정답: 독립사건이 아니다.**  
**풀이**  

사건 A: 영어시험 합격  
사건 B: 수학시험 합격

주어진 값:
- P(A) = 0.5
- P(B) = 0.4
- P(A ∩ B) = 0.25

두 사건이 **독립**이라면,
$$ P(A \cap B) = P(A)P(B) $$
이어야 한다.

우변 계산:
$$ P(A)P(B) = 0.5 \times 0.4 = 0.2 $$

하지만 실제 P(A ∩ B) = 0.25 이므로,
$$ 0.25 \neq 0.2 $$

따라서 두 사건은 **독립이 아니다**.

```python
p_A = 0.5
p_B = 0.4
p_A_and_B_given = 0.25
p_A_times_B = p_A * p_B
p_A_times_B, p_A_and_B_given
```

##### 6. 여사건을 이용한 확률 계산 (적어도 1번 성공)
동전을 3번 던질 때, **적어도 1번 앞면**이 나올 확률을 구하시오.  
가능하면 **여사건(한 번도 앞면이 나오지 않을 사건)**을 이용하여 계산해 보시오.

**정답: 7/8 (0.875)**  
**풀이 (여사건 이용)**  

사건 A: "적어도 한 번 앞면이 나오는 사건"  
여사건 Aᶜ: "한 번도 앞면이 나오지 않는 사건(= 모두 뒷면)"

동전 3번 던질 때 모두 뒷면이 나올 확률:
$$ P(모두 뒷면) = (1/2)^3 = 1/8 $$

따라서 적어도 한 번 앞면이 나올 확률은
$$ P(A) = 1 - P(A^c) = 1 - 1/8 = 7/8 = 0.875 $$

```python
prob_all_tail = (1/2)**3
prob_at_least_one_head = 1 - prob_all_tail
prob_all_tail, prob_at_least_one_head
```

##### 7. 조건부 확률
전체 학생 중 40%는 **프로그래밍을 좋아하고**, 그중 60%는 **파이썬을 좋아한다**고 한다.  

전체 학생 중에서 임의로 한 명을 뽑았을 때, 그 학생이 **파이썬을 좋아할 확률**을 구하시오.  

**정답: 0.24**  

**풀이**

사건 A: "프로그래밍을 좋아한다"  
사건 B: "파이썬을 좋아한다"  
여기서 파이썬을 좋아하는 학생은 **프로그래밍을 좋아하는 학생들 중에서 60%**이다.

주어진 값:
- P(A) = 0.4
- P(B | A) = 0.6

우리가 구해야 할 것은 P(B) = P(A)P(B|A)

$$ P(B) = 0.4 \ 0.6 = 0.24 $$

따라서 전체 학생 중 **파이썬을 좋아할 확률은 0.24**이다.

```python
p_program = 0.4
p_python_given_program = 0.6
p_python = p_program * p_python_given_program
p_python
```

##### 경우의 수를 이용한 확률 (주사위 2개)

##### 9. 사건의 독립성 판단
두 사건 A, B에 대해 다음과 같은 정보가 주어진다.
- P(A) = 0.6  
- P(B) = 0.5  
- P(A ∩ B) = 0.3  

다음 보기 중에서 옳은 것을 고르시오.

1) A와 B는 서로 배반사건이다.  
2) A와 B는 독립사건이다.  
3) A와 B는 독립사건이 아니다.

**정답: 2) A와 B는 독립사건이다.**  

**풀이**  

주어진 값:
- P(A) = 0.6
- P(B) = 0.5
- P(A ∩ B) = 0.3

두 사건이 **독립**이라고 하려면
$$ P(A \cap B) = P(A)P(B) $$
가 성립해야 한다.

우변 계산:
$$ P(A)P(B) = 0.6 \times 0.5 = 0.3 $$

주어진 P(A ∩ B)가 0.3이므로,
$$ P(A \cap B) = P(A)P(B) $$
가 성립한다.

따라서 A와 B는 **독립사건**이고, 정답은 **2번**이다.  

- 1) 배반사건이라면 P(A ∩ B) = 0 이어야 하지만, 여기서는 0.3이므로 틀림.  
- 3) 독립이 아니라고 보기에는 조건을 만족하고 있으므로 틀림.  

```python
p_A = 0.6
p_B = 0.5
p_A_and_B = 0.3
p_A_times_B = p_A * p_B
p_A_and_B, p_A_times_B
```
