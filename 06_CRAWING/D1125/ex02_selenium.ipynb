### 【 동적 웹 페이지 크롤링 】
- [예제] 검색창에 글자 입력 + 검색 버튼 클릭

[1] 모듈 로딩 및 웹 브라우저 실행 <hr>

```python
## - 모듈 로딩 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import selenium 

print(f'selenium v.{selenium.__version__}')
```

```python
## - 브라우저 실행
driver = webdriver.Chrome()
driver.get("https://www.google.com")
```

[2] 검색어 입력 및 실행<hr>
- send_keys() : 입력/엔터 등 보내기
- find_element(By.XXX, "값") : 1개 찾기
- find_elements(By.XXX, "값") : 여러개 찾기

```python
## (1) 검색창 찾기 (name속성의 값 'q')
search_box = driver.find_element(By.NAME, "q")

## (2) 검색어 입력
search_box.send_keys("파이썬 Selenium 크롤링")

## (3) 엔터로 검색 실행
search_box.send_keys(Keys.RETURN)

## (4) 결과가 로딩될 때까지 잠깐 대기 
time.sleep(30)

## (5) 검색 결과 제목들 가져오기
titles = driver.find_elements(By.CSS_SELECTOR, "h3")
for t in titles[:5]:
    print(t.text)

## (6) 브라우저 닫기
driver.quit()
```
