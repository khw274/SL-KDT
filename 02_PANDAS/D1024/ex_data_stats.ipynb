[3] 통계 메서드 사용 <hr>


```python
#_____________________________________________________
# => 컬럼별 통계값 계산 및 출력
#_____________________________________________________
# -> 컬럼별 계산
for col in mpgDF.columns:
    # 컬럼명 출력 및 숫자형 여부 확인
    print(col, mpgDF[col].dtype != object, sep=" ")
    print(f"[{col}] {mpgDF[col].dtype.name:8} --- {mpgDF[col].dtype != object}")

    # 숫자형인 경우 통계값 출력
    if mpgDF[col].dtype != object:
        print(f"mean : {mpgDF[col].mean(numeric_only=True)}")
        print(f"min : {mpgDF[col].min(numeric_only=True)}")
        print(f"max : {mpgDF[col].max(numeric_only=True)}")
        print(f"median : {mpgDF[col].median(numeric_only=True)}")
        # mode는 numeric_only 옵션 없이 사용
        print(f"mode : {mpgDF[col].mode()[0]}")  # 최빈값 중 첫 번째 값 출력
        print(f"sum : {mpgDF[col].sum(numeric_only=True)}")
```

```python
#_____________________________________________________
# => 컬럼별 통계값 계산 및 출력
#_____________________________________________________
# -> 컬럼별 계산
for col in mpgDF.columns:
    # 컬럼명 출력 및 숫자형 여부 확인
    print(col, mpgDF[col].dtype != object, sep=" ")
    print(f"[{col}] {mpgDF[col].dtype.name:8} --- {mpgDF[col].dtype != object}")

    # 숫자형인 경우 통계값 출력
    if mpgDF[col].dtype != object:
        print(f"mean : {mpgDF[col].mean(numeric_only=True)}")
        print(f"min : {mpgDF[col].min(numeric_only=True)}")
        print(f"max : {mpgDF[col].max(numeric_only=True)}")
        print(f"median : {mpgDF[col].median(numeric_only=True)}")
        # mode는 numeric_only 옵션 없이 사용
        print(f"mode : {mpgDF[col].mode()[0]}")  # 최빈값 중 첫 번째 값 출력
        print(f"sum : {mpgDF[col].sum(numeric_only=True)}")
```

```python
#_____________________________________________________
# => 데이터프레임 전체 통계값 계산 및 출력
#_____________________________________________________
for func in [mpgDF.mean, mpgDF.std]:
    print(f'{func}----------------------------')
    display(mpgDF.mean(numeric_only = True))
```

[4] 컬럼과 컬럼의 관계성 <hr>
- 상관계수: corr() 메서드
    * 값의 범위: -1 ~ -1
    * 종류: 음의 상관관계, 양의 상관관계

```python
mpgDF.corr(numeric_only=True)
```

```python
# 연비mpg 컬럼과 다른 컬럼들의 상관계수
mpg_corrSR = mpgDF.corr(numeric_only=True).loc['cylinders':, 'mpg']

mpg_corrSR.sort_values()
```

```python
# ==> mpg 컬럼과 관련성이 높은 컬럼은 weight, displacement, cylinders [음의 상관관계]
#                                 model_year, origin              [양의 상관관계]
#     종합해서 봤을 때 weight, displacement, cylinders이 연비와 높은 관계성이 있음
```
