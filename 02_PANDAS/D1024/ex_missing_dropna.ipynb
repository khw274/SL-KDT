### [데이터 전처리 - 결측치 삭제]
- 결측치(Missing Value)
    * 데이터/값이 존재하지 않는 것
- 결측치 체크
    * isna(): 결측치 여부에 따라 True/False 반환
- 결측치 처리
    * 삭제: dropna()
    * 치환: fillna()

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# 모듈 로딩
import pandas as pd
import numpy as np

# DF 생성
df = pd.DataFrame( { 'age' : [ 5, 6, np.nan ],
                     'born': [ pd.NaT, pd.Timestamp('1939-05-27'), pd.Timestamp('1940-04-25')],
                     'name': ['Alfred', 'Batman', ''],
                     'toy' : [None, 'Batmobile', 'Joker']})

# DF 출력
display(df)

# 기본 요약 정보
df.info()
```

[2] 결측치 체크: isna() / isnull() <hr>

```python
# 결측치 체크
display(df.isna())

# True => 1, False => 0: 계산 진행
print(f"컬럼별 결측치 수 \n{df.isna().sum()}")
```

[3] 결측치 삭제 : dropna() <hr>

```python
# 결측치 존재하는 행 제거 : 행 단위로 검사 진행해서 1개라도 결측치 있으면 삭제
df.dropna()
```

```python
# 결측치 존재하는 열 제거 => axis=1 또는 'columns' 설정

df.dropna(axis = 1)
# df.dropna(axis = 'columns')
```

```python
# 결측치 존재하는 행/열 제거
# how : 제거 방식/기준 설정
#       'all' : 모든 속성이 결측치인 행
#       'any' : 1개 이상 속성이 결측치인 행 [기본]

df.dropna(how='all')
```

```python
# 결측치 존재하는 행/열 제거
# thresh : 결측치가 아닌 최소 데이터 개수 설정
#          how와 함께 사용 불가

df.dropna(thresh=2)
```

```python
df.dropna(thresh=3)
```

```python
df.dropna(thresh=4)
```

```python
# subset : 특정 컬럼 또는 행만 결측치 검사 설정
# df.dropna(subset = ['name'])  # <= name 컬럼이 결측치면 행 삭제
df.dropna(subset = ['name'])    # <= born 컬럼이 결측치인 행 삭제
```

```python
# subset : 특정 컬럼 또는 행만 결측치 검사 설정

df.dropna(subset = ['name', 'toy'])
```

```python
# subset : 특정 컬럼 또는 행만 결측치 검사 설정

df.dropna(subset = ['name', 'toy'], how='all')
```
