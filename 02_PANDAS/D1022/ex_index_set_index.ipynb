[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# 모듈 로딩
import pandas as pd
import numpy as np      # 데이터가 없다 => NaN, nan, Na, na 표기용
```

```python
# 결측치(Missing Value) 표기법
# pd.NA : Not available
# pd.NaT : Not a Time
# pd.nan : Not a Number

print(pd.NA, pd.NaT, np.nan)
```

```python
dataDF = pd.DataFrame( [[1, 22, 39, 4, 85],
                        [11,22, 53, 7,55]] ,
                        index=[100, 29],
                        columns=['A','D','E','B','C'])

display(dataDF)
```

```python
#________________________________________________________
# 행 인덱스 기준으로 정렬 : [기본값] 행방향 오름차순 정렬
#________________________________________________________
dataDF.sort_index()         
```

```python
#________________________________________________________
# 행 인덱스 기준으로 정렬 : [기본값] 행방향 내림차순 정렬
#________________________________________________________
dataDF.sort_index(ascending = False)         
```

```python
#________________________________________________________
# 열 인덱스/열 이름 기준으로 정렬 : [변경] 열방향 내림차순 정렬
#________________________________________________________
display(dataDF)

dataDF.sort_index(axis = 'columns',
                  ascending = False)  
```

```python
#________________________________________________________
# 열 인덱스/열 이름 기준으로 정렬 : [변경] 열방향 오름차순 정렬
#________________________________________________________
display(dataDF)

dataDF.sort_index(axis = 'columns')  
```
