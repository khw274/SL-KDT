문제 1. 아래 데이터를 저장 후 조건을 만족하는 코드를 작성하세요. <hr>

```python
import pandas as pd
```

```python
# (1-1) 종목명, PER, PBR 컬럼만 선택 후 DataFrame으로 저장하는 코드 작성하세요.

data = [["2차전지(생산)", "SK이노베이션", 10.19, 1.29], 
        ["해운", "팬오션", 21.23, 0.95], 
        ["시스템반도체", "티엘아이", 35.97, 1.12], 
        ["해운", "HMM", 21.52, 3.20], 
        ["시스템반도체", "아이에이", 37.32, 3.55], 
        ["2차전지(생산)", "LG화학", 83.06, 3.75]]

columns = ["테마", "종목명", "PER", "PBR"]

dataDF = pd.DataFrame(data, columns=columns)
display(dataDF)

bool_index = ["종목명", "PER", "PBR"]
ret_data = dataDF[bool_index]
display(ret_data)
```

```python
# (1-2) PER이 20미만, 50이상인 데이터만 저장하는 코드 작성하세요.

retDF = (dataDF['PER'] < 20) | (dataDF['PER'] >= 50)
ret_data = dataDF[retDF]
display(ret_data)
```

```python
# (1-3) 종목명을 행인덱스로 설정하는 코드 작성하세요.

dataDF = dataDF.set_index('종목명')
display(dataDF)
```

```python
# (1-4) PBR 컬럼의 데이터를 기준으로 내림차순 정렬해 주세요.
dataDF = dataDF.sort_values(by = 'PBR', ascending = False)
display(dataDF)
```

```python
# (1-5) PBR, PER 순서로 오름차순 정렬해 주세요.
dateDF = dataDF.sort_values(by = ['PBR', 'PER'], ascending = [True, True])
display(dataDF)
```

```python
# 문제 4. 아래 데이터를 저장 후 조건을 만족하는 코드를 작성하세요.
FILE_NAME = '../Data/vgsales.csv'

dataDF = pd.read_csv(FILE_NAME)
display(dataDF)
```

```python
# (4-1) 'Rank' 컬럼을 인덱스로 설정하면서 DataFrame으로 변환하세요.

dataDF = dataDF.set_index('Rank')
display(dataDF)
type(dataDF)
```

```python
# (4-2) 'Global_Sales' 컬럼을 기준으로 내림차순 정렬하세요.

dataDF = dataDF.sort_values(by='Global_Sales', ascending=False)
display(dataDF)
```

```python
# (4-3) 'Platform'과 'Global_Sales'를 기준으로 정렬하세요. 
#       'Platform'은 오름차순으로 정렬하고 'Global_Sales'는 내림차순으로 정렬

dataDF = dataDF.sort_values(by=['Platform', 'Global_Sales'], ascending=[True, False])
display(dataDF)
```

```python
# (4-4) 'Year' 컬럼의 타입을 int로 변경하세요.

# NaN 값을 0으로 대치시킨 후 int형 전환
dataDF['Year'] = dataDF['Year'].fillna(0)
dataDF['Year'] = dataDF['Year'].astype(int)
display(dataDF)
dataDF['Year'].dtype

# Int64는 NaN값 허용됨
```

```python
# (4-5) 게임 이름(Name)에 'Super Mario'가 포함된 게임들만 선택하세요. 

ret_data = dataDF["Name"].str.contains('Super Mario')
display(dataDF[ret_data])
```

문제 5. 아래 데이터를 저장 후 조건을 만족하는 코드를 작성하세요. <hr>


```python
FILE_NAME = '../Data/titanic.csv'

dataDF = pd.read_csv(FILE_NAME)
```

```python
# (5-1) 단순 조건에 맞는 데이터만 선택 
# ▶ 성별이 여자인 승객 행만 필터링.
ret_data = dataDF['sex'] == 'female'
display(dataDF[ret_data])

# ▶ pclass가 1등석인 승객만. 
ret_data = dataDF['pclass'] == 1
display(dataDF[ret_data])

# ▶ age가 10세 미만인 어린이만.
ret_data = dataDF['age'] < 10
display(dataDF[ret_data])

# ▶ fare가 100이상인 고가 요금 승객만.
ret_data = dataDF['fare'] >= 100
display(dataDF[ret_data])
```

```python
# (5-2) 아래 조합 조건에 맞는 데이터만 선택
# ▶ 1 또는 2등석인 생존자 데이터만
ret_data = (dataDF['pclass'] == 1) | (dataDF['pclass'] == 2)
display(dataDF[ret_data])

# ▶ 가족 동승 3명 이상) 승객 데이터만
ret_data = dataDF['sibsp'] + dataDF['parch'] >= 3
display(dataDF[ret_data])

# ▶ 혼자 탑승한 성인 남자 데이터만 
ret_data = (dataDF['alone'] == True) & (dataDF['adult_male'] == 'True') 
display(dataDF[ret_data])
```
