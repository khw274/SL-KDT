### [ 속성 .iloc[] / .loc[] ]
- .iloc[] : 위치 기반 인덱스만 사용 가능
- .loc[]  : 라벨/네임 인덱스만 사용 가능

[설정 인덱스가 숫자인 경우] <hr>

```python
import pandas as pd
```

```python
sr = pd.Series([99, 78], index = [1, 2])
display(sr)

print("1번 원소 선택")
print("99 => ", sr[1], sr.loc[1], sr.iloc[0])   # iloc만 인덱스가 바뀌지 않음

print("2번 원소 선택")
print("78 => ", sr[2], sr.loc[2], sr.iloc[1])
```

[설정인덱스가 글자인 경우] <hr>


```python
sr = pd.Series([99, 78], index = ['a', 'b'])
display(sr)

print("원소 선택")
print("99 => ", sr['a'], sr.loc['a'], sr.iloc[0])   # iloc만 인덱스가 바뀌지 않음
print("78 => ", sr['b'], sr.loc['b'], sr.iloc[1])
```
