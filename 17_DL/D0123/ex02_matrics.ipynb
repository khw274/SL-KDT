```python
## 모듈로딩
from sklearn.metrics import r2_score, mean_squared_error, root_mean_squared_error
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score

import torch

from torchmetrics.regression import *
from torchmetrics.classification import *
from torchmetrics import functional as fn
```

[1]회귀 성능지표<hr>

```python
#- 임시데이터
target = torch.tensor([3, -0.5, 2, 7])
preds  = torch.tensor([2.5, 0.0, 2, 8])
```

```python
#- 성능평가 
print("r2_score     :", r2_score(target, preds))
print("mse          :", mean_squared_error(target, preds))
print("rmse         :", root_mean_squared_error(target, preds))
```

```python
## ------------------------------------------------------------------
## [1-2] torchmetrics
## ------------------------------------------------------------------
## => 인스턴스 방식
print("r2_score     :", R2Score()(preds, target).item())
print("mse          :", MeanSquaredError()(preds, target).item())
print("rmse         :", MeanSquaredError(squared=False)(preds, target).item())

## => 함수 방식
print("r2_score     :", r2_score(preds, target))
print("mse          :", mean_squared_error(target, preds))
print("rmse         :", root_mean_squared_error(target,preds))
```

[2] 분류 성능지표<hr>

```python
#- 임시데이터
target = torch.tensor([1, 1, 1, 1, 1])
preds  = torch.tensor([1, 0, 0, 1, 0])


#- 성능평가 
print("accuracy         :", accuracy_score(target, preds))
print("f1_score         :", f1_score(target, preds))
print("precision_score  :", precision_score(target, preds))
print("recall_score     :", recall_score(target, preds))
```

```python
## ------------------------------------------------------------------
## [2-2] torchmetrics
## ------------------------------------------------------------------
## => 인스턴스 방식
print("F1Score           :", F1Score(task='binary', num_classes=2)(preds, target).item())
print("BinaryF1Score     :", BinaryF1Score()(preds, target).item())

print("Accuracy          :", Accuracy(task='binary', num_classes=2)(preds, target).item())
print("BinaryAccuracy    :", BinaryAccuracy()(preds, target).item())

print("Precision          :", Precision(task='binary', num_classes=2)(preds, target).item())
print("BinaryPrecision    :", BinaryPrecision()(preds, target).item())

print("Recall            :", Recall(task='binary', num_classes=2)(preds, target).item())
print("BinaryRecall      :", BinaryRecall()(preds, target).item())

## => 함수 방식
print("accuracy_score     :", fn.accuracy( preds, target, task='binary', num_classes=2 ).item())
print("r2_score           :", fn.f1_score(preds, target, task='binary', num_classes=2 ).item())
print("precision_score    :", fn.precision(preds, target, task='binary', num_classes=2 ).item())
print("recall_score       :", fn.recall(preds, target, task='binary', num_classes=2 ).item())
```
