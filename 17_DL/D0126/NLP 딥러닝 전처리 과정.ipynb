#### NLP 딥러닝 모델 전처리 과정
1. 텍스트 정제 (Text Cleaning): 소문자 변환, 특수문자·숫자·HTML 태그 제거
​
2. 토큰화 (Tokenization): 문장을 단어/형태소/서브워드 단위로 분할

3. 불용어 제거 (Stop Word Removal): "the", "는", "가" 등 의미 없는 단어 제거

4. 정규화 (Normalization): 축약어 확장, 띄어쓰기·이모지 정리
​
5. 어간/표제어 추출 (Stemming/Lemmatization): 단어를 기본형으로 변환
​
6. 벡터화 (Vectorization): 텍스트를 수치 벡터로 변환 (TF-IDF, Word2Vec, FastText)
​
7. 인코딩 (Encoding): 범주형 라벨을 수치로 변환 (One-Hot Encoding, Label Encoding)
​
8. 스케일링 (Scaling): 수치 데이터를 정규화/표준화 (Min-Max, StandardScaler)

9. 패딩 (Padding): 모든 시퀀스를 동일한 길이로 맞춤
​
