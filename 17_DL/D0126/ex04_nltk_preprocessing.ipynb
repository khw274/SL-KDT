```python
# ================================================================================
# [1-1] ëª¨ë“ˆ ë¡œë”©
# ================================================================================
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.tag import pos_tag
from nltk.stem import WordNetLemmatizer, PorterStemmer, LancasterStemmer

from collections import Counter
# ================================================================================
# [1-2] ë°ì´í„° ì¤€ë¹„ - Louis de Beauvoir ì „ê¸° í…ìŠ¤íŠ¸ ë¡œë“œ
# ================================================================================
with open("../DATA/Louis_Jean_Joseph.txt", mode='r') as f:
    allData = f.read()

print(f"allData => {len(allData)}ê°œ")
print(f"allData => {allData[:100]}")
```

[2] í…ìŠ¤íŠ¸ ë°ì´í„° ì „ì²˜ë¦¬<hr>
- ê¸°ë³¸ ì •ì œ : êµ¬ë‘ì , ì–¸ì–´ë³„ ë¶ˆìš©ì–´
- í† í°í™” ì§„í–‰ => í† í° ê¸°ë°˜ ì •ì œ í•„ìš”í•  ìˆ˜ ìˆìŒ

- ì˜ì–´ ê²½ìš° ëŒ€ì†Œë¬¸ì ì¼ì¹˜ 

```python
# ================================================================================
# [2-1] í† í°í™” ë° ì†Œë¬¸ìë¡œ í†µì¼
# ================================================================================
# ì†Œë¬¸ìë¡œ í†µì¼
allData = allData.lower()

# í† í°í™” - ë‹¨ì–´ ë‹¨ìœ„ë¡œ ë¶„ë¦¬
wordList = word_tokenize(allData)

print(f"ê¸°ë³¸ í† í° : {wordList[:20]}\n")
print(f"ê¸°ë³¸ í† í° : {len(wordList)}\n")
```

```python
# ================================================================================
# [2-2] í† í° ì •ì œ
# ================================================================================
puncList = list(punctuation)        # !@#$%^&*()
swList = stopwords.words("english") # the, is, are...

print(wordList[:20])
print(f"ê¸°ë³¸ í† í° : {len(wordList)}\n")

# - êµ¬ë‘ì  ì œê±°
wordList = [word for word in wordList if word not in puncList]
print(wordList[:20])
print(f"êµ¬ë‘ì  ì œê±° í›„ í† í° : {len(wordList)}\n")

# - ë¶ˆìš©ì–´ ì œê±°
wordList = [word for word in wordList if word not in swList]
print(wordList[:20])
print(f"ë¶ˆìš©ì–´ ì œê±° í›„ í† í° : {len(wordList)}\n")
```

```python
# ================================================================================
# [2-3] Stemming (ì–´ê·¼ ì¶”ì¶œ)
# ================================================================================
lanStem = LancasterStemmer()    # ê³µê²©ì 
portStem = PorterStemmer()      # ì •í™•ë„ ë†’ìŒ

for word in wordList[:20]:
    print(f"{word:<16} --> Lan : {lanStem.stem(word):<16} port : {portStem.stem(word):<16}")
```

```python
# ================================================================================
# [2-4] Lemmatization (í‘œì œì–´ ì¶”ì¶œ) : í‘œì œì–´ -> ì‚¬ì „ì— ê¸°ë³¸í˜•ìœ¼ë¡œ ì‹¤ë¦° ë‹¨ì–´
# ================================================================================
wordLem = WordNetLemmatizer()   # WordNet ì‚¬ì „ ê¸°ë°˜

# í’ˆì‚¬ 'v'(ë™ì‚¬)ë¡œ ê°•ì œ ë³€í™˜
for word in wordList[:20]:
    print(f"Pos = 'v' : {wordLem.lemmatize(word, pos='v'):<16}")
print('\n')

# 3ê°€ì§€ ë¹„êµ
for word in wordList[:20]:
    # 1. ì›ë³¸ 2. ê¸°ë³¸ (ëª…ì‚¬ 'n') 3. ë™ì‚¬ ('v')
    print(f"{word:<16} --> No Tag : {wordLem.lemmatize(word):<16} Tag(v) : {wordLem.lemmatize(word, pos='v'):<16}")
```

```python
# ================================================================================
# [2-5] í’ˆì‚¬íƒœê¹…ì‘ì—… : ì •í™•í•˜ê²Œ í˜•íƒœì†Œ ë¶„ë¦¬ ìœ„í•´ì„œ ì§„í–‰ => í† í° ë‹¨ìœ„ë¡œ ì§„í–‰
# ================================================================================
text = ' '.join(wordList[:1000])  # wordList ê¸°ë°˜ ìƒ˜í”Œ

tk_pos = pos_tag(word_tokenize(text), tagset="universal")
tkList = []

for tk, pos in tk_pos:
    if pos in ['NOUN', 'VERB']:
        tkList.append((tk, pos)) 

print("ì „ì²´ íƒœê·¸ ìˆ˜:", len(tk_pos))
print("ì„ íƒëœ íƒœê·¸ (NOUN/VERB) :", tkList[:20])  
print(f"ì´ ì„ íƒ ìˆ˜: {len(tkList)}ê°œ")
```

[3] ë‹¨ì–´/ì–´íœ˜ ì‚¬ì „ ìƒì„± <HR>

```python
# ================================================================================
# [3-1] í† í°ì˜ ë¹ˆë„ìˆ˜ ì²´í¬ ë° ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
# ================================================================================
allTKDict = dict(Counter(tk for tk, pos in tkList))  # [ìˆ˜ì •1] íŠœí”Œ â†’ í† í°ë§Œ
rAllTKList = sorted(allTKDict.items(), key=lambda x: x[1], reverse=True)

print("ìƒìœ„ 10ê°œ í† í°:")
for token, freq in rAllTKList[:10]:  
    pos = next((p for t, p in tkList if t == token), 'NOUN')
    print(f"{token:<15} (POS: {pos}): {freq:>4}íšŒ")


# ================================================================================
# [3-2] ë‹¨ì–´/ì–´íœ˜ ì‚¬ì „ ìƒì„±
# ================================================================================
vocab = {"oov": 0}
for tk, _ in rAllTKList:
    vocab[tk] = len(vocab)

print(f"\nì–´íœ˜ ì‚¬ì „ í¬ê¸°: {len(vocab)-1}ê°œ ë‹¨ì–´")
print("ì²˜ìŒ 10ê°œ:", dict(list(vocab.items())[:11]))
```

```python
# ì¸ì½”ë”©ëœ ë¬¸ì¥ ë¦¬ìŠ¤íŠ¸ ì €ì¥ìš© 
sentEncodeList = []

# ë¬¸ì¥ë³„ í† í°ì„ ìˆ«ì/ì •ìˆ˜ ì¸ì½”ë”© ì§„í–‰
for tk, pos in tkList:  # íŠœí”Œ ì–¸íŒ¨í‚¹
    token_id = vocab.get(tk, 0)  # í† í°ë§Œ ì‚¬ìš©
    sentEncodeList.append([token_id])  # ë‹¨ì¼ í† í° ë¬¸ì¥

print("ì¸ì½”ë”© ê²°ê³¼ (ì²˜ìŒ 20ê°œ):")
print(sentEncodeList[:20])
print(f"ì´ ë¬¸ì¥ ìˆ˜: {len(sentEncodeList)}")
```
