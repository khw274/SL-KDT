### [ Python Class ]


```python
# ======================================================================================
# 클래스이름 : 사람 Person
# 클래스속성 : 이름, 성별, 나이
# 클래스기능 : 먹는다, 잔다
# ======================================================================================
class Person:
    # 인스턴스 생성 시 호출되는 메서드 -> 객체/인스턴스 생성 시 자동 호출: Callback
    def __init__(self, name_, gender_, age_):
        self.name = name_
        self.gender = gender_
        self.age = age_
    
    # 인스턴스 메서드 : 1번째 매개변수 self!
    def eat(self, food):
        print(f"{self.name}가 {food}을/를 먹는다")

    def sleep(self):
        print(f"{self.name}가 잔다")

    # 매직 함수 / 메서드 재정의 / 오버라이딩
    # => + 연산자 
    def __add__(self, obj):
        print("__add__")
        return self.age + obj.age
    
    def __len__(self):
        print("지금까지 살아온 날 수")
        return self.age * 365
    
    # => 객체/인스턴스변수명()
    def __call__(self, data):
        print('__call__')
        self.sleep()
```

```python
# 사람 인스턴스/객체 생성
p1 = Person('박혁거세', '남', 100)
p2 = Person('마징가'  , '여', 99)

# 속성 사용하기 
# - 변수명.속성명            => 현재 속성 읽기
# - 변수명.속성명 = 새로운 값 => 속성 변경
print("p1의 나이 : ", p1.age)

p2.gender = '남'

# 메서드 사용하기
# - 변수명.메서드이름()
p1.sleep()
p2.eat("막창")
```
