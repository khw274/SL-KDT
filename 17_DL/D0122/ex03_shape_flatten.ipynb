### [ nn.flatten 클래스 - 텐서 형태 변경 ]
- 형태 변경 관련 Layer 클래스 중 1개
- 이미지 다룰 때 2차원/3차원 -> 1차원으로 변환에 많이 사용됨

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# ======================================================================================
# [1-1] 모듈 로딩
# ======================================================================================
import pandas as pd

import torch
import torch.nn as nn
import torch.nn.functional as F     # 신경망 함수들(AF, LF, MF) 모듈

from torchinfo import summary
from collections import OrderedDict

# ======================================================================================
# [1-2] 데이터 준비
# ======================================================================================
dataDF = pd.read_csv("../DATA/study_score_multi.csv")
display(dataDF.head())
display(dataDF.info())
```

[2] Flatten Layer 활용한 형태 변경 <hr>

```python
# 데이터 텐서 생성
dataTS = torch.tensor([[[11, 22, 33], [12, 23, 34]], [[44, 55, 66], [45, 56, 67]]],
                      dtype=torch.float32)

print(dataTS.size(), dataTS.ndim)
print(dataTS)
```

```python
# FC Layer에 전달
fcLay = nn.Linear(6, 1)
fcLay
```

```python
# FC Layer <-- Tensor 전달
# 3D (샘플수, 행, 열) -> 2D (샘플수, 피쳐수)
print(f"전 : {dataTS.shape}")

flatLA = nn.Flatten()
dataTS = flatLA(dataTS)

print(f"후 : {dataTS.shape}")

fcLay(dataTS)
```
