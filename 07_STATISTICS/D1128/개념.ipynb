## 기술통계
- 중심위치/대표값 : 
    - 평균         => 이상치 : 절사평균/가중평균 
    - 중앙값/중간값 => 오름차순 정렬 후 가운데 있는 값
    - 최빈값       => 수치형, 범주형

- 산포도(대표값 기준 흩어짐 정도) : 
    - 편차(데이터 - 평균) 모든 편차 합계 = 0
    - 분산(편차 제곱), 절대평균
    - 표준편차(편차의 제곱근)
    - 표준편차(편차 제곱근)
    - 범위(최대값 - 최소값)
 
- 분포:                          
    - 분위수 => 이분위수, 사분위수, 백분위수
    - 기준: 정규분포
    - 왜도(비대칭_왼쪽/오른쪽), 첨도(뽀족 정도)

- 관계성
    1. 주제에 해당하는 컬럼(속성, attribute, field, feature, 특성)과 관련 있는 컬럼 
    2. 유사한 컬럼들
    3. 컬럼들 타입 고려 => 수치형/범주형
