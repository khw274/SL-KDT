```python
## 모듈 로딩
import pandas as pd 
import numpy as np
import wquantiles as wq       
```

```python
## 임의의 데이터
dataA  = [70, 72, 75, 75, 76, 80, 82, 85, 88, 90]
dataSR = pd.Series(dataA)
```

[2] 데이터 정보확인<hr>

```python
## 기술통계값 확인 
dataSR.describe(percentiles=[0.1, 0.9])
```

```python
## ====================================================================
## 평균/분산/표준편차 : pandas와 numpy의 분산/표준편차 계산식의 차이로 값 다름
## ====================================================================
## => pandas 
mean_pd = dataSR.mean()
var_pd  = dataSR.var(ddof=1)
std_pd  = dataSR.std(ddof=1)
print(f'mean_pd : {mean_pd},  var_pd : {var_pd}, std_pd : {std_pd}')

## => numpy
mean_np = np.mean(dataSR)
var_np  = np.var(dataSR)
std_np  = np.std(dataSR)
print(f'mean_np : {mean_np},  var_np : {var_np}, std_np : {std_np}')
```
