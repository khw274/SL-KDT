```python
## [ìƒì„¸í•œ í’€ì´] ---------------------------------------------------------------------
## ì£¼ì‚¬ìœ„ë¥¼ êµ´ë ¤ 3ì˜ ë°°ìˆ˜ê°€ ë‚˜ì˜¬ ë•Œë¥¼ ì„±ê³µìœ¼ë¡œ í•˜ëŠ” ì‹¤í—˜(ì„±ê³µì˜ í™•ë¥ ì´ 1/3 ì¸ ë² ë¥´ëˆ„ì´ ì‹œí–‰)ì„ 
## 3ë²ˆ ë…ë¦½ìœ¼ë¡œ ë°˜ë³µí•´ì„œ ì‹¤í—˜í•  ë•Œì˜ ì„±ê³µ íšŸìˆ˜ì˜ í™•ë¥ ì§ˆëŸ‰í•¨ìˆ˜ë¥¼ êµ¬í•´ ë´…ì‹œë‹¤.
##
## â˜… ëª©í‘œ: ì£¼ì–´ì§„ í™•ë¥ ë¡œ í™•ë¥  ë¶„í¬ í•¨ìˆ˜ë¥¼ ìœ ë„
## -> ëª…í™•í•œ ì£¼ì–´ì§„ í™•ë¥  : ëª…í™•í•œ í™•ë¥ (p=1/3) ì£¼ì–´ì§!! ì¦‰, ì´ë¡ ì ì¸ í™•ë¥ ê°’(ëª¨ìˆ˜), ì¦‰ ì‹¤í—˜ ê·œì¹™
##    1/3ì€ ì£¼ì‚¬ìœ„ì˜ ë©´ì´ 6ê°œì´ê³  ê·¸ì¤‘ 3ì˜ ë°°ìˆ˜ê°€ 2ê°œ(3, 6)ë¼ëŠ” ìˆ˜í•™ì  ì‚¬ì‹¤ì—ì„œ ì •ì˜ëœ ê°’
## -> ì‹œë„ íšŸìˆ˜ : 3ë²ˆ
## -> ì„±ê³µ í™•ë¥  : 1/3
## -> ì„±ê³µ íšŸìˆ˜ : 0ë²ˆ, 1ë²ˆ, 2ë²ˆ, 3ë²ˆ => 4ê°€ì§€
## -> ì„±ê³µ íšŸìˆ˜ë³„ í™•ë¥ ì„ ëª¨ë‘ ê³„ì‚°í•´ì£¼ëŠ” í•¨ìˆ˜ => í™•ë¥ ì§ˆëŸ‰í•¨ìˆ˜(pmf)
from scipy.stats import binom

n = 3      # ì‹œí–‰ íšŸìˆ˜
p = 1/3    # ì„±ê³µ í™•ë¥  
k = 1      # ì„±ê³µí•œ íšŸìˆ˜ (0,1,2,3) ì¦‰, í™•ë¥ ë³€ìˆ˜
```

```python
from scipy.stats import binom

n = 8               # ì‹œí–‰ íšŸìˆ˜ (ì˜ˆ: 8ë¬¸ì œ)
p = 1/8             # ì„±ê³µí™•ë¥  (ì˜ˆ: í•œ ë¬¸ì œ ë§ì¶œ í™•ë¥ )
k = [0,1,2,3]       # í™•ë¥ ë³€ìˆ˜ Xê°€ ì·¨í•  íŠ¹ì • ì„±ê³µ íšŸìˆ˜ ê°’

prob = binom.pmf(k, n, p)   # P(X = 1)
print(prob, sum(prob))
```

```python
## [ìƒì„¸í•œ í’€ì´] ---------------------------------------------------------------------
## ì–´ë–¤ ë¬¸ì œë¥¼ ë§ì¶œ í™•ë¥ ì´ ì¼ì •í•œ ìƒí™©ì—ì„œ,  
## í•™ìƒì´ 8ë¬¸ì œë¥¼ ì°ì–´ì„œ í’€ì—ˆê³  ê·¸ì¤‘ 3ë¬¸ì œê°€ ë§ì•˜ë‹¤.  ==> ì‹¤ì œë¡œ ê´€ì¸¡ëœ ê²°ê³¼(ë°ì´í„°) ì¦‰, ê´€ì¸¡ê°’
## ì´ë•Œ í•œ ë¬¸ì œë¥¼ ë§ì¶œ **ì„±ê³µí™•ë¥  pì˜ ì¶”ì •ê°’**ì„ êµ¬í•˜ì‹œì˜¤.  
##
## â˜… ëª©í‘œ: ì´ë¯¸ ë²Œì–´ì§„ ê²°ê³¼ë¥¼ ë°”íƒ•ìœ¼ë¡œ ì„±ê³µ í™•ë¥ ì„ ì—­ìœ¼ë¡œ ì¶”ì •
## -> ê´€ì¸¡ê°’/ê²°ê³¼ : í•™ìƒì´ 8ë¬¸ì œë¥¼ ì°ì–´ì„œ í’€ì—ˆê³  ê·¸ì¤‘ 3ë¬¸ì œê°€ ë§ì•˜ë‹¤. ìš°ì—°í•œ ê²°ê³¼
## -> ì„±ê³µ  íšŸìˆ˜ : 0ê°œ, 1ê°œ, ..., 8ê°œ => 8ê°€ì§€
## -> ì„±ê³µí™•ë¥ ê°’ ì¶”ì²­ ê³µì‹ : ì„±ê³µíšŸìˆ˜/ì „ì²´ìˆ˜ = 3/8
p = 3/8
print(f'ì„±ê³µí™•ë¥  p =>{p}')
```

```python
## ì„±ê³µ í™•ë¥  p=3/8ì¸ ë² ë¥´ëˆ„ì´ ì‹œí–‰ì„ 8ë²ˆ ë…ë¦½ì ìœ¼ë¡œ ë°˜ë³µí–ˆì„ ë•Œ 
## í•œë²ˆ ì„±ê³µí•˜ëŠ” í™•ë¥ ì§ˆëŸ‰í•¨ìˆ˜ë¡œ í™•ë¥ ë¶„í¬ë¥¼ êµ¬í•˜ì„¸ìš”.
## => ê¸°ë³¸ì ìœ¼ë¡œ 8ë²ˆ ì‹œí–‰í–ˆì„ ë•Œì˜ ê²°ê³¼
## => ì„±ê³µ/ì‹¤íŒ¨ : ì´í•­ë¶„í¬
## => ê³µì‹ : P(X=k) =  C(n, k) * p^k * (1-p)^(n-k)  # ì´í•­ ë¶„í¬ í™•ë¥  ì§ˆëŸ‰ í•¨ìˆ˜ ìˆ˜ì‹
from scipy.stats import binom

n = 8               # ì‹œí–‰ íšŸìˆ˜ (ì˜ˆ: 8ë¬¸ì œ)
p = 3/8             # ì„±ê³µ í™•ë¥  (ì˜ˆ: 3ë¬¸ì œ ë§ì¶˜ ì„±ê³µí™œë¥ )
k = range(9)        # í™•ë¥  ë³€ìˆ˜ Xê°€ ì·¨í•  íŠ¹ì • ì„±ê³µ íšŸìˆ˜ ê°’

prob = binom.pmf(k, n, p)   # P(X = 1)
print(prob, sum(prob))
print(f'8ê°œ ì¤‘ 1ê°œ ë§ì¶œ í™•ë¥ : {prob[1]}')
```
