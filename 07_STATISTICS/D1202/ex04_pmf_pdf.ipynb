#### 【 확률분포관련 함수들 】
- 확률질량함수(PMF:Probability Mass Function)
- 확률밀도함수(PDF:Probability Density Function)
- 누적분포함수(CDF:Cumulative Distribution Function)

<pre>
<hr>
* 확률질량함수(PMF: Probability Mass Function)
  - 이산확률변수가 특정 값(x)을 가질 확률을 직접 나타내는 함수
  - 정의 : P(X=x)=PMF(x)
  - 특징 : 모든 확률은 0이상, 전체 확률 합은 1
  - 막대그래프
  - 베르누이, 이항, 포아송
<hr>
</pre>

```python
## 모듈 로딩
import numpy as np
import matplotlib.pyplot as plt

# 데이터 및 확률질량계산
x   = np.arange(1, 7)
pmf = [1/6] * 6

# 시각화
plt.bar(x, pmf)
plt.title("PMF of Fair Dice")
plt.xlabel("x")
plt.ylabel("P(X=x)")
plt.show()
```

<pre><hr>
* 확률밀도함수(PDF: Probability Density Function)
  - 연속확률변수가 특정 구간 안에 존재할 확률의 밀도(density) 나타내는 함수
  - 실수(real number)처럼 값이 연속적인 경우
  - 특징 : 특정 값의 확률 P(X=a) = 0
  - 곡선(curve) 그래프
  - 정규, 지수, t-분포에 사용
<hr></pre>

```python
## 모듈 로딩
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 300)
pdf = norm.pdf(x, 0, 1)

plt.plot(x, pdf)
plt.title("PDF of Standard Normal Distribution")
plt.xlabel("x")
plt.ylabel("density")
plt.grid(True)
plt.show()
```

<hr>[ PMF & PDF 비교]<hr>

```python
## 모듈 로딩
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

## 시각화
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

##-> PMF (이산)
x = np.arange(1, 7)
pmf = [1/6] * 6
axes[0].bar(x, pmf)
axes[0].set_title("PMF (Discrete)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("P(X=x)")

##-> PDF (연속)
x2 = np.linspace(-4, 4, 300)
pdf = norm.pdf(x2, 0, 1)
axes[1].plot(x2, pdf)
axes[1].set_title("PDF (Continuous)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("density")
axes[1].grid(True)

plt.show()
```

<pre><hr>
* 누적분포함수(CDF:Cumulative Distribution Function)
  - 확률 이해하는 핵심 도구이며, 이산·연속 분포 모두를 한 번에 다룰 수 있는 함수
  - 확률변수 X가 x 이하의 값을 가질 누적 확률
  - F(x)=P(X≤x)

  - 성질
    -> 항상 0에서 시작, 1에서 끝남
    -> 증가하는 함수 (줄지 않음)
    -> 이산은 계단형, 연속은 매끈한 곡선
    -> PDF는 CDF를 미분한 것
    -> CDF는 해석하기 가장 쉬운 확률 요약 도구

  - 적용
    -> 지금까지 값들이 얼마나 누적되었는가?
    -> x보다 작거나 같은 값이 나올 확률은?
    -> PMF나 PDF를 일정 지점까지 합치거나(이산) 적분(연속)한 값
<hr></pre>

[1]CDF 계산 & 그래프 그리기<hr>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 7)
pmf = np.ones(6) / 6
cdf = np.cumsum(pmf)

plt.step(x, cdf, where='mid')
plt.title("CDF of Fair Dice")
plt.xlabel("x")
plt.ylabel("P(X ≤ x)")
plt.grid(True)
plt.show()
```

[2] 연속 분포 CDF (정규분포)<hr>

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 300)
cdf = norm.cdf(x, 0, 1)

plt.plot(x, cdf)
plt.title("Standard Normal CDF")
plt.xlabel("x")
plt.ylabel("P(X ≤ x)")
plt.grid(True)
plt.show()
```

[3] CDF로 특정 확률 계산하기<hr>

```python
from scipy.stats import norm

# P(0<X<1.96)=F(1.96)−F(0)
result = norm.cdf(1.96) - norm.cdf(0)
print(result)   # 약 0.475
```
