#### 【 연속확률분포 (Continuous Distribution) 】
- 개념 및 특징
    - 확률 변수가 특정 구간 내의 모든 실수를 가질 수 있는 분포
    - 확률 밀도 함수(PDF) 사용하여 특정 구간의 넓이를 계산하여 확률을 구함

    - 대표분포 : 균일분포, 정규분포, 지수분포, t-분포, 카이제곱분포, F-분포

[1] 균일분포 (Uniform Distribution)<hr>
- 구간 [a, b]에서 모든 값이 똑같이 나올 만큼 공평한 분포
- 예시
    - 0초~10초 사이에서 랜덤 시간 한 번 뽑기
    - 0~1 사이에서 random.random()

```python
# 모듈로딩
import numpy as np
import matplotlib.pyplot as plt

# [0, 1] 구간의 균일분포 샘플
samples = np.random.uniform(0, 1, size=10000)

# 시각화
plt.hist(samples, bins=30, density=True)
plt.title("Uniform(0,1) Distribution")
plt.xlabel("x")
plt.ylabel("density")
plt.grid(True)
plt.show()
```

[2] 정규분포 (Normal Distribution) <hr>
- 가장 중요한 연속분포. 벨 모양(종 모양).
- 파라미터: 평균 μ, 표준편차 σ
- 많은 자연/사회/측정 데이터가 근사적으로 따름
- 중심극한정리 때문에 통계·ML에서 핵심

```python
# 모듈로딩
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, pi, exp

# 표준정규분포 N(0,1)의 pdf 함수
def normal_pdf(x, mu=0, sigma=1):
    return (1/(sqrt(2*pi)*sigma)) * np.exp(-0.5 * ((x-mu)/sigma)**2)

# 1) PDF 곡선 그리기
x = np.linspace(-4, 4, 300)
y = normal_pdf(x, mu=0, sigma=1)

plt.plot(x, y)
plt.title("Normal PDF (μ=0, σ=1)")
plt.xlabel("x")
plt.ylabel("density")
plt.grid(True)
plt.show()

# 2) 샘플 히스토그램 + 이론 pdf 비교
samples = np.random.normal(loc=0, scale=1, size=10000)

plt.hist(samples, bins=40, density=True, alpha=0.7, label="samples")
plt.plot(x, y, label="theoretical pdf")
plt.title("Normal Distribution: Samples vs PDF")
plt.xlabel("x")
plt.ylabel("density")
plt.legend()
plt.grid(True)
plt.show()
```

[3] 지수 분포(Exponential Distribution)<hr>
- 다음 사건까지 걸리는 **시간을 모델링**할 때 자주 쓰는 분포

- 파라미터: λ (rate, 단위시간당 발생률)
- 예
    * 다음 전화가 올 때까지 시간
    * 다음 고장까지 시간
    * 다음 손님이 올 때까지 대기 시간
    * 포아송 과정의 간격

```python
# 모듈로딩
import numpy as np
import matplotlib.pyplot as plt

# 변수 및 데이터 생성
lam = 1.5  # λ
samples = np.random.exponential(scale=1/lam, size=10000)

# 히스토그램
plt.hist(samples, bins=40, density=True, alpha=0.7)

# 이론 pdf 시각화 
x = np.linspace(0, 5, 300)
pdf = lam * np.exp(-lam * x)
plt.plot(x, pdf)

plt.title(f"Exponential Distribution (λ={lam})")
plt.xlabel("x")
plt.ylabel("density")
plt.grid(True)
plt.show()
```

```python
# - 연속확률분포는 값이 연속적인 확률변수의 분포
# - PDF로 밀도를 표현
# - 확률은 구간의 면적(적분) 계산
# - 균일·정규·지수는 가장 자주 쓰이는 대표 3형제
```
