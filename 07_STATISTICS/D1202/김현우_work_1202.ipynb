20. 한 카페에 손님이 평균적으로 1분당 2명 방문한다고 한다.  
    이 카페에 1분 동안 손님이 0명 방문할 확률을 푸아송분포로 구하시오.  
단.from scipy.stats import poisson

- 공식: P(X = k) = (e^(-λ) × λ^k) / k!

- 여기서:
    - λ = 2 (평균)
    - k = 0 (구하는 값)
    - e ≈ 2.718

- 계산:

- P(X = 0) = (e^(-2) × 2^0) / 0!
    - = e^(-2) × 1 / 1
    - = e^(-2)
    - ≈ 0.1353
    - ≈ 13.53%
    - 답: e^(-2) ≈ 0.1353 (약 13.53%)

- Python 코드

    - from scipy.stats import poisson

- ### 포아송 분포, λ=2
    - poi = poisson(mu=2)

- ### P(X=0) 계산
    - prob = poi.pmf(0)
    - print(f"확률: {prob:.4f}")      # 0.1353
    - print(f"백분율: {prob*100:.2f}%")  # 13.53%
