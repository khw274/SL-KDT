```python
## ===============================================================
## [문제 1] 개념 비교
## ===============================================================
## 일반 평균, 절사 평균, 가중 평균의 공통점과 차이점을 간단히 서술하세요.
## - 세 가지 평균의 공통점은 모두 데이터의 대표값을 나타내기 위한 추정값입니다.
## - 세 가지 평균의 차이점은 사용 상황에 따라서 평균을 구하는 방법이 다른 것인데
##   ㄴ 일반 평균은 모든 데이터를 더해서 개수로 나누고,
##   ㄴ 절사 평균은 극단값을 제거 후에 평균을 계산, 
##   ㄴ 가중 평균은 각 데이터에서 중요도, 즉 가중치를 곱한 후 가중치의 총합으로 나눈 값입니다.
## ===============================================================
## 다음 상황에서 어떤 평균을 쓰는 것이 적절한지 고르고 이유를 쓰세요.

## (1) 극단값이 포함된 연봉 데이터의 대표값
##     ㄴ ✔️ 극단값이 포함되어 있다면 일반 평균값이 부정확할 수 있기 때문에 절사 평균으로 극단값을 제거하고 계산하면 적절한 평균이 나올 것 같습니다.
 
## (2) 각기 다른 표본 수를 가진 실험 그룹들의 평균 반응값 통합
##     ㄴ ✔️ 표본의 수가 극단적이라면 표본 수를 가중치로 잡고 가중평균을 구해야 적절할 거 같습니다.

## (3) 시험 점수 중 최근 시험에 더 높은 비중을 주고 싶을 때
##     ㄴ ✔️ 더 높은 비중을 주고 싶다면 가중평균을 사용해서 최근 시험에 높은 가중치를 두고 평균을 구하는 것이 적절할 거 같습니다.
```

```python
## ===============================================================
## [문제 2] 반별 평균을 통합할 때 – 가중평균
## ===============================================================
## A반과 B반의 시험 결과가 다음과 같다.
##
## A반: 학생 수 10명, 평균 90점
##
## B반: 학생 수 40명, 평균 70점

## 전체 50명에 대한 단순 평균을 잘못 계산한 경우 (반 평균의 평균)와
## 올바른 전체 평균(가중평균) 을 계산하세요.

## ✔️ 단순 평균: (90 + 70) / 2 = 80점
## ✔️ 가중 평균: (10 * 90 + 40 * 70) / 50 = 74점
```

```python
## ===============================================================
## [문제 3] 절사 평균 계산
## ===============================================================
## 데이터는 아래와 같습니다. 
x = [10, 12, 11, 9, 13, 200]

## 일반 평균을 구하시오.
sum = 0

for i in x:    
    sum += i

print(f"✔️ 일반 평균: {sum / len(x)}")  # ✔️ 42.5점

## 가장 작은 값과 가장 큰 값(10, 200)을 하나씩 제거한 후 절사 평균을 구하시오.
print(f"✔️ 절사 평균: {np.mean(sorted(x)[1:-1])}")

## 왜 절사 평균이 일반 평균보다 극단값에 덜 민감한지 설명하시오.
## ✔️ 절사 평균은 극단값을 모두 제거하고 평균을 구하기 때문에 극단값의 영향이 없어 덜 민감하다.
```

```python
## ===============================================================
## [문제 4] 가중 평균 계산
## ===============================================================
## 어떤 회사의 고객 만족도 조사는 다음과 같습니다. 
##  그룹	응답자 수	평균 점수
##  VIP	       20	    4.8
##  일반	  200	    4.2
##
## 전체 고객의 가중 평균 만족도 점수를 계산하시오.
print(f"✔️ 가중 평균 만족도 점수: {(20 * 4.8 + 200 * 4.2) / 220}")

## 단순히 (4.8 + 4.2) / 2 로 계산하는 것이 왜 잘못되었는지 설명하시오.
## ✔️ 일반 고객이 전체 응답자 수 중 91%를 차지하고 있기 때문에 일반 평균으로는 정확한 평균을 구할 수 없음.
```

```python
## ===============================================================
##  [문제 5] Pandas로 그룹별 가중평균 계산 
## ===============================================================
## 다음과 같은 DataFrame이 있습니다.

df = pd.DataFrame({
    "group":  ["A", "A", "A", "B", "B"],
    "score":  [80, 90, 70, 60, 100],
    "weight": [1, 2, 1, 3, 1]
})


## group별 단순 평균(score의 평균)을 구하시오.
print(f"✔️ group별 단순 평균(score의 평균): \n{df.groupby('group')['score'].mean()}\n")

## group별 가중평균(weight를 가중치로 사용)을 구하는 Pandas 코드를 작성하시오.
print("✔️ group별 가중평균(weight를 가중치로 사용):")
print(df.groupby('group').apply(lambda x : np.average(x["score"], weights = x['weight'])))
```

```python
## ===============================================================
##  [ 문제 6 ] 분산과 표준편차 비교
## ===============================================================
## 다음은 어느 반 학생 10명의 수학 시험 점수이다.
## A반: [70, 72, 75, 75, 76, 80, 82, 85, 88, 90]
## B반: [40, 55, 60, 70, 75, 90, 95, 98, 100, 100]

A = [70, 72, 75, 75, 76, 80, 82, 85, 88, 90]
B = [40, 55, 60, 70, 75, 90, 95, 98, 100, 100]

## 두 반의 분산(sample variance)을 각각 구하시오.
print(np.var(A, ddof=1))              # A반 표본 분산
print(np.var(B, ddof=1), '\n')      # B반 표본 분산

## 두 반의 표준편차를 비교하고, 어느 반의 점수 분포가 더 불안정한지 설명하시오.
print(np.std(A, ddof=1))              # A반 표준 편차
print(np.std(B, ddof=1))            # B반 표준 편차

## ✔️ A반의 표준편차는 6.85, B반의 표준편차는 21.52로 B반이 약 3배 이상 불안정함.
## A반 편차 범위 = 90-70 = 20, 즉 70 ~ 90점
## B반 편차 범위 = 100-40 = 60, 즉 40 ~ 100점(불안정)
```
