### 【 확률 (probability) 】
- 표본을 여러 번 뽑아서 표본평균이 어떻게 분포하는지 확인 
    * 모집단에서 여러 번 표본을 뽑으면
    * 표본 평균들이 하나의 분포를 이룸
    * 분포는 정규분포에 가까워짐(중심 극한 정리:CLT, Central Limit Theorem)
    - ★ 추론통계의 핵심!

[1] 동전 던지기 시뮬레이션<hr>

```python
## ------------------------------------------------------
## 모듈 로딩
## ------------------------------------------------------
import random
```

```python
## 재현성 고정
random.seed(42)
```

```python
## ------------------------------------------------------
## 함수기능 : 동전 한 번 던지기: 'H' = 앞면, 'T' = 뒷면
## 함수이름 : toss_coin
## 반환결과 : 앞/뒤 반환
## ------------------------------------------------------
def toss_coin():
    return random.choice(['H', 'T'])

# 동전 여러 번 던져보기
num_trials  = 1000  # 시행 횟수
heads_count = 0

## num_trials => 10, 100, 1000, 10000 으로 변경
##               점점 0.5 근처로 수렴하는 모습을 볼 수 있음.
for num_trials in [10, 100, 1000, 10000]:
    for _ in range(num_trials):
        if toss_coin() == 'H':
            heads_count += 1

    prob_heads = heads_count / num_trials
    print(f"동전 {num_trials}번 던져서 앞면이 나온 비율: {prob_heads:.3f}")

```

[2] 주사위 시뮬레이션: “짝수 나올 확률” 확인  <hr>

```python
## ------------------------------------------------------
## 함수기능 : 주사위 한 번 던지기
## 함수이름 : roll_die
## 반환결과 : 1~6 중 하나 반환
## ------------------------------------------------------
def roll_die():
    return random.randint(1, 6)

## 시행 횟수 및 짝수 출력 횟수 
num_trials = [10, 100, 1000, 10000]
even_count = 0

for num in num_trials:
    for _ in range(num):
        if roll_die() % 2 == 0:  # 짝수인지 확인
            even_count += 1

    prob_even = even_count / num
    print(f"주사위 {num}번 던져서 짝수 나온 비율: {prob_even:.3f}")
```
