#### 【 Kendall Tau Correlation Coefficient 】

- 설문 예제 : 만족도 순위 vs 재구매 의사 순위 관계 분석
- 10명의 응답자에게
    * 제품 만족도 (1~5점, 5가 매우 만족)
    * 재구매 의사 (1~5점, 5가 매우 재구매 의사 높음)
- 값을 그대로 순위/서열(ordinal) 데이터로 보고 켄달 τ 계산

```python
## ======================================================
## 모듈 로딩
## ======================================================
import numpy as np
from scipy.stats import kendalltau
```

```python
## ======================================================
## [예제] 만족도 vs 재구매 의사
## ======================================================
## 10명의 응답자 (id 1~10)
respondents = np.arange(1, 11)


## 예시 데이터 (1~5점 Likert 척도라고 가정)
## 만족도: 1(매우 불만) ~ 5(매우 만족)
satisfaction = np.array([5, 4, 3, 2, 4, 5, 3, 2, 1, 4])

## 재구매 의사: 1(전혀 없음) ~ 5(매우 높음)
repurchase = np.array([5, 3, 3, 1, 4, 5, 2, 2, 1, 4])

print("응답자 ID      :", respondents)
print("만족도(1~5)    :", satisfaction)  
print("재구매 의사(1~5):", repurchase)

tau, _ = kendalltau(satisfaction, repurchase)

print("\n[설문 예제] Kendall's tau =", tau)
```
