- 확률변수 => 
    * 시행 결과로 올 수 있는 값들 : 표본공간. 집합으로 표현 => 전집합{1, 2, 3, 4, 5, 6}
    * 사건(Event)의 결과로 나온 값은 표본공간/전집합의 원소들 부분집합{1}, {4}, {2} => 근원집합
    * 합집합/교집합/여집합 => 합사건, 교사건, 여사건, ...

- 확률변수: 숫자화(정수, 실수) 변환한 것 : 평균, 분산...
    * 동전 던지기       => 앞면, 뒷면 ===> 1,0
    * 크리스마스 이브 눈 => 온다/안온다 ===> 1,0
    * 주사위 던기지      => 1,2,3,4,5,6 ===> 1,2,...,6    

- 확률분포: 확률변수의 확률값에 대한 구성 분포
    * 확률변수 정수 => 이산확률분포
    * 확률변수 실수 => 연속확률분포

- 확률분포함수: 
    * 이산확률분포함수 => 확률질량함수(PMF)  
    * 연속확률분포함수 => 확률밀도함수(PDF)
    * 누적분포함수 (CDF)
