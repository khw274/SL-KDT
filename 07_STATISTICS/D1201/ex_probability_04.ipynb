### 【 표본추출 】
- 비복원 추출 : 동일 데이터/값 나올 수 없음
- 복원 추출 : 동일 데이터/값 나올 수 있음.

```python
## ------------------------------------------------------
## 모듈 로딩
## ------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
import random 
```

```python
## 복원 추출: 같은 값이 여러 번 나올 수 있음
def sample_with_replacement(population, k=5):
    return [random.choice(population) for _ in range(k)]

## 비복원 추출: 중복 없이 선택
def sample_without_replacement(population, k=5):
    return random.sample(population, k=k)
```

```python
## -------------------------------------------
## 샘플링 진행
## -------------------------------------------
population = list(range(1, 11))  # 1~10
print("\n[복원 / 비복원 추출 데모]")
print("모집단:", population)

## 복원 & 비복원 샘플링 
sample_with    = sample_with_replacement(population, k=10)
sample_without = sample_without_replacement(population, k=5)

print("  복원 추출 (10개):", sample_with)
print("  비복원 추출 (5개):", sample_without)

if np is not None:
    pop_np              = np.arange(1, 11)
    sample_with_np      = np.random.choice(pop_np, size=5, replace=True)  # 복원
    sample_without_np   = np.random.choice(pop_np, size=5, replace=False) # 비복원
    print("  복 원 추출 (numpy, 5개):", sample_with_np)
    print("  비복원 추출 (numpy, 5개):", sample_without_np)
else:
    print("  (numpy가 없어서 numpy 예제는 건너뜁니다.)")

```
