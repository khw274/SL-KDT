### [데이터 시각화 - Line Graph]
- 시간 흐름에 따른 변화를 시각화
- 다양한 선 스타일과 마커를 활용
- 함수: matplotlib.pyplot.plot()

```python
# [1] 모듈 로딩
import matplotlib.pyplot as plt 
import koreanize_matplotlib     
```

```python
# [2] 기본 선그래프
# - 선 그래프 인스턴스 생성: 기본값으로 선 그래프 생성
#          plot(y)      : y축 데이터의 수 만큼 x축 데이터 생성(0 ~ N-1개)
line = plt.plot([10, 20, 30, 40, 50])

# - 화면에 그래프 출력
plt.show()

# - Line 인스턴스 속성 정보 읽기 
print(line)
```

```python
# [3] 사용자 설정 선 그래프
# - 선 그래프 인스턴스 생성: 기본값으로 선 그래프 생성
# plot(x, y):    x축 데이터,                 y축 데이터,            fmt 선스타일 설정 => 색상 마커스타일 선스타일
line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r')
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], '.g')

# - 화면에 그래프 출력
plt.show()
```

```python
# [4] 사용자 설정 선 그래프
# - 선 그래프 인스턴스 생성: 기본값으로 선 그래프 생성
# plot(x, y):    x축 데이터, y축 데이터, 
#                fmt 선스타일 설정 => 색상 마커스타일 선스타일
#                **kwargs        => Line2D의 속성들 설정
#                 x축 데이터                  y축 데이터             fmt  **kwargs
line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r', markersize=20)
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], '.g', markeredgecolor='yellow')

# - 화면에 그래프 출력
plt.show()
```

```python
# [5] 사용자 설정 선 그래프
# - 선 그래프 인스턴스 생성: 기본값으로 선 그래프 생성
# plot(x, y):    x축 데이터, y축 데이터, 
#                fmt 선스타일 설정 => 색상 마커스타일 선스타일
#                **kwargs        => Line2D의 속성들 설정
# Line2D 속성값들 저장 => Line2D 속성명:값
propertys = {'ms':20,
             'mec':'yellow',
             'ls':'-.'}

#                 x축 데이터                  y축 데이터             fmt  **kwargs
line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r', markersize=20)
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], '.g', **propertys)

# - 화면에 그래프 출력
plt.show()
```

```python
# [6] 그래프의 인스턴스 속성 읽기 : 객체변수명.get_속성명()
line1_, line2_ = line1[0], line2[0]

print(f'Line1 그래프 속성들 ---------\n')
print(f'마커 사이즈: {line1_.get_ms()}')
print(f'마커 색상: {line1_.get_mfc()}')

print(f'Line2 그래프 속성들 ---------\n')
print(f'마커 사이즈: {line2_.get_ms()}')
print(f'마커 색상: {line2_.get_mfc()}')
print(f'마커 테두리: {line2_.get_mec()}')
print(f'선 스타일: {line2_.get_ls()}')
print(f'데이터: {line2_.get_data()}')
```
