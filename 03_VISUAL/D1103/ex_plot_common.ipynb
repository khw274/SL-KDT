### [데이터 시각화 - 공통사항]
- 모든 그래프에 공통적으로 들어가는 요소들
- 그래프 제목, x축 라벨, y축 라벨
- 한 axes에 여러개 그래프가 존재하는 경우 식별용 범례(legend)
- x, y축에 눈금(ticks) 2종류 즉, major ticks / minor ticks

```python
# [1] 모듈 로딩
import matplotlib.pyplot as plt 
import koreanize_matplotlib     
```

```python
# [2] 사용자 설정 선 그래프
# - 선 그래프 인스턴스 생성: 기본값으로 선 그래프 생성

#                 x축 데이터                  y축 데이터             fmt  **kwargs
line = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50])

# 그래프 공통 요소 설정 => 각 요소별 인스턴스 반환
font_dict = {'size':20, 
             'color':'red', 
             'weight':'bold'}

label_dict = {'color':'blue',
              'fontsize':20}

# -> 그래프와 타이틀 간격 pad 매개변수 설정
# -> 그래프에 대한 타이틀 위치 loc 매개변수 설정: 'left', 'center', 'right'
plt.title('[선 그래프]', loc='right', pad=20, **font_dict)

# - 눈금과 라벨들의 간격 조절 labelpad 매개변수 설정
# - 라벨들의 위치 loc 매개변수 설정 : x, y 라벨에 따라 값 다름
plt.xlabel('종류',  labelpad=20, loc='right', **label_dict)  # labelpad= : 간격조절, # loc= : 라벨 위치 설정// 행: 'left', 'center', 'right'
                                                            #                                              열: 'top', 'center', 'bottom
plt.ylabel('값', label_dict)

# - 화면에 그래프 출력
plt.show()
```

```python
# [3] 사용자 설정 선 그래프
# - 선 그래프 인스턴스 생성: 기본값으로 선 그래프 생성
# plot(x, y):    x축 데이터, y축 데이터, 
#                fmt 선스타일 설정 => 색상 마커스타일 선스타일
#                **kwargs        => Line2D의 속성들 설정
# 여러개 그래프 식별을 위한 범례 설정
# -> 방법1: 그래프에 label속성 설정 + plt.legend()

# label 매개변수 설정 => 범례로 사용
#                 x축 데이터                  y축 데이터             fmt  **kwargs
line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r', label='line1')
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], '.g', label='line2')

# 범례 화면 출력
plt.legend()

# - 화면에 그래프 출력
plt.show()
```

```python
# [4] 사용자 설정 선 그래프
# - 선 그래프 인스턴스 생성: 기본값으로 선 그래프 생성
# plot(x, y):    x축 데이터, y축 데이터, 
#                fmt 선스타일 설정 => 색상 마커스타일 선스타일
#                **kwargs        => Line2D의 속성들 설정
# 여러개 그래프 식별을 위한 범례 설정
# -> 방법2: plt.legend(labels=[])

line1 = plt.plot(['a', 'b', 'c', 'd', 'e'], [10, 20, 30, 40, 50], '+r', label='line1')
line2 = plt.plot(['a', 'b', 'c', 'd', 'e'], [15, 25, 35, 45, 55], '.g', label='line2')

# 범례 화면 출력. labels 매개변수: 그래프 수 만큼 리스트 담아 설정
# 범례 배치 방식 매개변수 ncols=1 설정: ncols=2
# 범례 출력 위치 매개변수 loc=str, int 설정 [기본값: loc='best']
# 범례 라벨 색상 매개변수 labelcolor='색상' 설정
font_dict = {'fontsize':15, 
             'labelcolor':'blue',
             'shadow':True,
             'title':'다양한 선그래프'}

plt.legend(labels=['L1', 'L2'], ncols=2, loc=4, **font_dict)
# plt.legend(labels=['L1', 'L2'], ncols=2, loc='upper right', **font_dict)
# plt.legend(labels=['L1', 'L2'], ncols=2, loc='lower left', **font_dict)

# - 화면에 그래프 출력
plt.show()
```
