### [선 그래프]
- 시간의 흐름에 따른 추세/변화/트렌드 표현

[1] 모듈 로딩 및 데이터 준비 <hr>

```python
# 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# 데이터 준비
data = {'A':[10, 20, 30, 40, 50], 'B':[15, 25, 35, 45, 55]}

dataDF=pd.DataFrame(data)
display(dataDF)
```

[2] A에 따른 B에 변화 시각화 <hr>

```python
# ---------------------------------------
# 선그래프 함수: plt.plot()
# ---------------------------------------
plt.plot(dataDF['A'], dataDF['B'])

# 그래프 기본 설정
plt.xlabel('A')
plt.ylabel('B')
plt.title('[A에 따른 B의 변화]')

# 그래프 화면 출력
plt.show()
```

```python
# ---------------------------------------
# 선그래프 함수: plt.plot(y, fmt)
# ---------------------------------------
plt.plot(dataDF['A'], 'b-')         # 'b-': blue 실선
plt.plot(dataDF['A']+10, 'r-')
plt.plot(dataDF['A']-20, 'r--')     # 'r--': red 점선
plt.plot(dataDF['A']-30, 'g^')      # 'g^': green 삼각형마크, 선X 
plt.plot(dataDF['A']-5, 'm^-')      # 'g^': mageta 삼각형마크, 실선

# 그래프 기본 설정
plt.xlabel('A')
plt.title('[A 변화]')

# 그래프 화면 출력
plt.show()
```

```python
# ---------------------------------------
# 선그래프 함수: plt.plot(y, fmt, label)
# ---------------------------------------
plt.plot(dataDF['A'], 'b-', label='A')         # 'b-': blue 실선
plt.plot(dataDF['A']+10, 'r-', label='A+10')
plt.plot(dataDF['A']-10, 'g^', label='A-10')      # 'g^': green 삼각형마크, 선X 
plt.plot(dataDF['A']-5, 'm^-', label='A-05')      # 'g^': mageta 삼각형마크, 실선

# 그래프 기본 설정
plt.xlabel('A')
plt.title('[A 변화]')
plt.legend()

# 그래프 화면 출력
plt.show()
```
