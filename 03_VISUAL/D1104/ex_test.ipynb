# [데이터 시각화]
- 데이터셋: iris.csv
- 해결문제: 품종별 시각화에 적합한 특성/속성/컬럼 선정
- 출력결과: 선정된 특성을 기반으로 품종 분류한 것 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt 
import koreanize_matplotlib
```

```python
FILE_NAME= '../DATA/iris.csv'
# sepal length	꽃받침 길이
# sepal width	꽃받침 폭 
# petal length	꽃잎 길이 
# petal width	꽃잎 폭

dataDF = pd.read_csv(FILE_NAME)
display(dataDF.head())

dataDF = dataDF.drop(['petal.length', 'petal.width'], axis=1)
display(dataDF.head())
```

```python
fig = plt.figure(figsize=(10, 6))
# plt.figure()

# 색상 설정
iris = {'Setosa':'red', 'Versicolor':'green', 'Virginica':'blue'}
display(iris.items())

# colors = ['red', 'green', 'blue']
# columns = ['Setosa', 'Versicolor', 'Virginica']

for variety, color in iris.items():
    # print(dataDF['variety'] == variety)

    varity_data = dataDF[dataDF['variety'] == variety]
    # print(varity_data)

    plt.scatter(varity_data['sepal.length'], varity_data['sepal.width'])

plt.title('iris')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(iris.keys())

plt.show()
```
