### [Matplotlib의 Pie 그래프]
- 전체 대비 항목에 대한 비중/비율 시각화 그래프
- 데이터 제공 시 자동 계산 수행

```python
import pandas as pd
import koreanize_matplotlib
import matplotlib.pyplot as plt
```

```python
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

wedgeprops={'width': 0.7,'edgecolor': 'white','linewidth': 5}

sizes = [30, 20, 25, 21] 

plt.pie(sizes,
        labels=labels, 
        colors=["#3bf251", "#4a58f1", "#ffcc4c", "#ff598b"],
        autopct="%1.1f%%",
        startangle=260, counterclock=False,
        pctdistance=1.3,
        labeldistance=0.6,
        wedgeprops=wedgeprops,
        explode = [0.2, 0, 0, 0])

# 현재 그래프의 figure, axes
fig = plt.gcf()
axec = plt.gca()

fig.set_facecolor('lightyellow')

plt.legend(loc = 'upper right', bbox_to_anchor=(1.1, 1.1))
plt.title('[항목별 비율]')
plt.show()
```
