### 02. 시리즈 생성(051 ~ 060) <hr>

```python
# 모듈 임포트

import pandas as pd
from pandas import Series
```

```python
#시리즈의 생성 (리스트)

data = [100, 200, 300]
s = pd.Series(data)
print(s)
```

```python
#시리즈의 생성 (인덱스 지정)

data = [100, 200, 300]
index = ['철수', '영희', '아름']
s = pd.Series(data, index)
print(s)
```

```python
#시리즈의 생성 (딕셔너리)

data = [100, 200, 300]
s = pd.Series(data)
s.index = ['철수', '영희', '아름']
print(s)
```

```python
# 시리즈와 딕셔너리

data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = pd.Series(data, index)

print(s.to_dict())
```

```python
s.to_excel('data.xlsx')
```

```python
# 시리즈와 엑셀

s.to_excel('data.xlsx', header=None)
```

```python
#시리즈와 엑셀

pd.read_excel('data.xlsx')
```

### 03. 시리즈 데이터 타입(061 ~ 070) <hr>

```python
# 시리즈 데이터 타입
data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = pd.Series(data, index)

print(s.dtype)
```

```python
# 시리즈 데이터

data = [  10,  "가" ]
s = pd.Series( data )

print(s.dtype)
```

```python
# 시리즈 인덱스의 데이터 타입
data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = pd.Series(data, index)

print(s.index.dtype)
```

```python
# 시리즈 이름과 인덱스 이름
data = {
    "2020-12-11": 73400,
    "2020-12-10": 72900,
    "2020-12-09": 73900,
    "2020-12-08": 71700,
    "2020-12-07": 72900
}

s = Series(data)
print(s)
s.name = '종가'
print(s)
s.index.name = "날짜"
print(s)
```

```python
# to_dict

data = {"apple": 1500, "banana": 3000}
s = pd.Series(data)
s
s.to_dict()
```

```python
# 인덱스 이름

data = {"apple": 1500, "banana": 3000}
s = pd.Series(data)
s
s.index.name = 'icecream'
s
```

```python
# 시리즈 이름

s = pd.Series(data=[1, 2, 3])
s.name = 'number'
s
```

```python
# NaN
data = [1.0, 2.0, None]
s = Series(data)
s
# s.loc[2] = 'NaN'
s
```

### 04 시리즈 인덱싱과 슬라이싱(071 ~ 080) <hr>

```python
# 시리즈 타입

s = pd.Series([1, 2, 'python'])
print(s)
print(s.dtype)
```

```python
# 시리즈 타입 변경

s = pd.Series(['1', '2', '3'])
s.astype('int32')
```

```python
# 시리즈 데이터 타입 설정

data = {'BTC':50000,
        'XRP':1,
        'ETH':4000}

s = Series(data, dtype='float')
s
```

```python
# 시리즈 인덱싱 (대괄호 사용하기)

data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = Series(data, index)
print(s['영희'])
```

```python
# 시리즈 인덱싱 (iloc 속성 사용하기)

data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = Series(data, index)

print(s.iloc[1])
```

```python
# 시리즈 인덱싱 (loc 속성 사용하기)

data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = Series(data, index)

print(s.loc['영희'])
```

```python
# 시리즈 인덱싱 (인덱스 사용하기)

data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = Series(data, index)

print(s.영희)
```

```python
# 시리즈 슬라이싱

data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = Series(data, index)

print(s.iloc[0:2])
print(s.iloc[ :2])
print(s.loc["철수":"영희"])
print(s[0:2])
print(s[ :2])
print(s["철수":"영희"])
print(s.values[:2])
```

```python
# 시리즈 슬라이싱

data = [100, 200, 300, 400, 500, 600]
index = ["08/01", "08/02", "08/03", "08/04", "08/05", "08/06"]
s = Series(data, index)

print(s.iloc[ : :2])
print(s[ : :2])

print(s.loc["08/01":"08/06":2])
print(s["08/01":"08/06":2])

print(s.iloc[[0, 2, 4]])
```

### 05. 시리즈 수정, 추가 및 삭제(081 ~ 090) <hr>

```python
# 시리즈 추가

data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = Series(data, index)

s['순이'] = 400
s
```

```python
# 시리즈 삭제

data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = Series(data, index)
s

s.drop('철수', inplace=True)
s
```

```python
# unique

data = {
    "삼성전자": "전기,전자",
    "LG전자": "전기,전자",
    "현대차": "운수장비",
    "NAVER": "서비스업",
    "카카오": "서비스업"
}

s = Series(data)
s

print(s.unique())
```
