```python
# 모듈 및 데이터 로딩

import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
import pandas as pd

FILE_NAME = '../SL_KDT_10/08_ML/Data/iris.csv'
dataDF = pd.read_csv(FILE_NAME)

display(dataDF)
```

```python
# 막대 그래프 - seaborn 추천

sns.barplot(data=dataDF, x='variety', hue='variety', y='sepal.length', errorbar=None, width=0.5)
plt.title("품종별 sepal 길이")
```

```python
sns.heatmap(data=dataDF.iloc[:3, :-1].corr(), annot=True, cmap='coolwarm', fmt='.1f', square=True, cbar=True)
plt.show()
```

```python
sns.scatterplot(data=dataDF, x='sepal.length', y='sepal.width', style='variety')
```

```python
sns.boxplot(data=dataDF)
```

```python
sns.histplot(data=dataDF, x='sepal.length', bins=20, hue='variety')
```

```python
sns.lineplot(data=dataDF)
```

```python
plt.pie(dataDF['sepal.length'], labels=dataDF['variety'], autopct='%.1f%%')
plt.show()
```

```python
# 여러 카테고리별로 동일한 그래프 그리기
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# variety 고유값으로 반복(품종별 개수)
varieties = dataDF['variety'].value_counts().index.tolist()
for idx, category in enumerate(varieties):
    subset = dataDF[dataDF['variety'] == category]
    sns.histplot(data=subset, x='sepal.length', ax=axes[idx])
    axes[idx].set_title(f'{category} (n={len(subset)})')
    
plt.tight_layout()
plt.show()
```
